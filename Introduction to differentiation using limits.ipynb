# Understanding Limits and Derivatives Using the Diagram

Differentiation is the process of finding the derivative of a function, which represents the rate of change of the function's value with respect to its input. One way to understand differentiation is through the concept of limits, as illustrated in the provided diagram.

## Diagram Explanation

The diagram shows a function $f(x)$ and how the difference quotient approximates the slope of the tangent line at a point as the interval $h$ approaches zero.



### Key Points in the Diagram

- **Point $(x, f(x))$**: This is a point on the function where we want to find the slope of the tangent line.
- **Point $(x+h, f(x+h))$**: This is another point on the function, $h$ units away from $x$.
- **Vertical Distances**: $f(x)$ and $f(x+h)$ represent the function's values at $x$ and $x+h$, respectively.
- **Horizontal Distance**: The distance between $x$ and $x+h$ is $h$.
- **Difference Quotient**: The slope of the secant line connecting $(x, f(x))$ and $(x+h, f(x+h))$ is given by the difference quotient:

  $$
  \frac{{f(x+h) - f(x)}}{h}
  $$

### Limit Concept

As $h$ approaches zero, the secant line becomes closer to the tangent line at $x$. The limit of the difference quotient as $h$ approaches zero gives the slope of the tangent line, which is the derivative of the function at $x$:

$$
f'(x) = \lim_{{h \to 0}} \frac{{f(x+h) - f(x)}}{h}
$$

### Step-by-Step Illustration

1. **Initial Secant Line**:
   - The blue line in the diagram represents the secant line connecting $(x, f(x))$ and $(x+h, f(x+h))$.
   - The slope of this secant line is the average rate of change of the function over the interval $[x, x+h]$.

2. **Approaching the Tangent Line**:
   - As $h$ decreases, the point $(x+h, f(x+h))$ moves closer to $(x, f(x))$.
   - The secant line starts to approximate the tangent line more closely.

3. **Limit Process**:
   - When $h$ is infinitesimally small, the secant line becomes the tangent line.
   - The slope of this tangent line is the derivative of $f(x)$ at $x$.

### Example with the Diagram

Consider a function $f(x)$ and let $a$ be the point at which we want to find the derivative. The diagram shows:

- The initial secant line (blue) between $(x, f(x))$ and $(x+h, f(x+h))$.
- As $h$ approaches zero, $(x+h, f(x+h))$ moves closer to $(x, f(x))$.
- The difference quotient $\frac{{f(x+h) - f(x)}}{h}$ gives the slope of the secant line.
- The limit of the difference quotient as $h$ approaches zero gives the slope of the tangent line at $x$, which is the derivative $f'(x)$.

In mathematical terms:

$$
f'(x) = \lim_{{h \to 0}} \frac{{f(x+h) - f(x)}}{h}
$$



### Visual Understanding

The green dashed line in the diagram represents the tangent line at $(x, f(x))$. The tangent line is the best linear approximation of the function at the point $x$. As $h$ becomes smaller, the secant line (blue) approaches this tangent line, and the difference quotient approaches the derivative.

## Conclusion

The diagram visually demonstrates how the concept of limits is used to define the derivative of a function. By understanding the difference quotient and its limit, we can find the slope of the tangent line to the function at any point, which is the essence of differentiation.



# Approximating Slopes of a Graph Using Limits

To understand how the slope of a function is approximated using limits, we can consider the function $$y = -\frac{x^2}{2} + 2x + 2$$ The provided diagrams show how the slope of the secant line approaches the slope of the tangent line as the points become closer.

## Step-by-Step Explanation

### Initial Secant Line

Consider two points on the curve: $(0.61, 3.034)$ and $(1.91, 3.996)$. The slope of the secant line between these points is calculated as follows:



1. **Difference in $y$-coordinates**:
   $$
   \Delta y = f(1.91) - f(0.61) = 3.996 - 3.034 = 0.962
   $$

2. **Difference in $x$-coordinates**:
   $$
   \Delta x = 1.91 - 0.61 = 1.3
   $$

3. **Slope of the secant line**:
   $$
   m = \frac{\Delta y}{\Delta x} = \frac{0.962}{1.3} \approx 0.74
   $$

### Closer Points

Now consider two points that are closer: $(0.61, 3.034)$ and $(1.26, 3.726)$. The slope of the secant line between these points is calculated as follows:



1. **Difference in $y$-coordinates**:
   $$
   \Delta y = f(1.26) - f(0.61) = 3.726 - 3.034 = 0.692
   $$

2. **Difference in $x$-coordinates**:
   $$
   \Delta x = 1.26 - 0.61 = 0.65
   $$

3. **Slope of the secant line**:
   $$
   m = \frac{\Delta y}{\Delta x} = \frac{0.692}{0.65} \approx 1.0646
   $$

### Approaching the Tangent Line

As the two points on the curve become closer, the slope of the secant line approaches the slope of the tangent line at the point of interest. The tangent line represents the instantaneous rate of change of the function at that point.



# Estimating the Slope at a Point

To estimate the slope of the function $y = 0.5x^2 - 3x + 3$ at the point $x = 3$, we can use the concept of the derivative. The slope at a point is given by the derivative of the function evaluated at that point. We can approximate this by choosing points very close to $x = 3$, such as $x = 3$ and $x = 3.1$.



## Function Definition

The function we are dealing with is:

$$
y = 0.5x^2 - 3x + 3
$$

## Steps to Estimate the Slope

1. Calculate the value of the function at $x = 3$.
2. Calculate the value of the function at $x = 3.1$.
3. Use these values to estimate the slope using the difference quotient formula.

### Step 1: Calculate $y$ at $x = 3$

First, substitute $x = 3$ into the function:

$$
y(3) = 0.5(3)^2 - 3(3) + 3
$$

Calculate the value:

$$
y(3) = 0.5 \cdot 9 - 9 + 3 = 4.5 - 9 + 3 = -1.5
$$

### Step 2: Calculate $y$ at $x = 3.1$

Next, substitute $x = 3.1$ into the function:

$$
y(3.1) = 0.5(3.1)^2 - 3(3.1) + 3
$$

Calculate the value:

$$
y(3.1) = 0.5 \cdot 9.61 - 9.3 + 3 = 4.805 - 9.3 + 3 = -1.495
$$

### Step 3: Estimate the Slope

The slope can be estimated using the difference quotient:

$$
\text{slope} \approx \frac{y(3.1) - y(3)}{3.1 - 3}
$$

Substitute the calculated values:

$$
\text{slope} \approx \frac{-1.495 - (-1.5)}{3.1 - 3} = \frac{-1.495 + 1.5}{0.1} = \frac{0.005}{0.1} = 0.05
$$

### Conclusion

Using the difference quotient, we estimate that the slope of the function $y = 0.5x^2 - 3x + 3$ at $x = 3$ is approximately $0.05$. This method provides an approximation of the derivative at the specified point by evaluating the function at points close to $x = 3$. As can be seen from the above graph, the slope is 0 since it is the minimum point in the quadratic equation.


# Estimating the Slope at a Point

To estimate the slope of the function $y = 0.5x^2 - 3x + 3$ at the point $x = 4$, we can use the concept of the derivative. The slope at a point is given by the derivative of the function evaluated at that point. We can approximate this by choosing points very close to $x = 4$, such as $x = 4$ and $x = 4.1$.

$$
y(4) = 0.5(4)^2 - 3(4) + 3

$$

### Step 2: Calculate $y$ at $x = 4.1$

Next, substitute $x = 4.1$ into the function:

$$
y(4.1) = 0.5(4.1)^2 - 3(4.1) + 3

$$
y(4.1) = 0.5 \cdot 16.81 - 12.3 + 3 = 8.405 - 12.3 + 3 = -0.895

$$
\text{slope} \approx \frac{y(4.1) - y(4)}{4.1 - 4}

$$
\text{slope} \approx \frac{-0.895 - (-1)}{4.1 - 4} = \frac{-0.895 + 1}{0.1} = \frac{0.105}{0.1} \approx 1.05

### Conclusion

Using the difference quotient, we estimate that the slope of the function $y = 0.5x^2 - 3x + 3$ at $x = 4$ is approximately $1.05$. This method provides an approximation of the derivative at the specified point by evaluating the function at points close to $x = 4$.
