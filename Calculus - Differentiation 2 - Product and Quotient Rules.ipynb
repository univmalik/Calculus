# 2.4 The Product and Quotient Rules

In this section, we explore how to differentiate products and quotients of functions.

## Derivative of a Product

The derivative of a product of two functions is not the product of their derivatives. For example, let's consider two functions:

- $f(x) = x$
- $g(x) = x^2$

The product of their derivatives is $f'(x)g'(x) = (1)(2x) = 2x$, but the product of the functions is $h(x) = f(x)g(x) = x^3$, and the derivative of this product is $h'(x) = 3x^2$.

### The Product Rule

The correct way to find the derivative of a product of two functions is provided by the Product Rule, formally stated as:

**Theorem 2.4.1 (The Product Rule):** If $f$ and $g$ are differentiable at $x$, then the product $f \cdot g$ is also differentiable at $x$, and its derivative is given by:

$$
\frac{d}{dx}[f(x)g(x)] = f(x) \frac{d}{dx}[g(x)] + g(x) \frac{d}{dx}[f(x)]
$$

Alternatively, this can be expressed as:

$$
(f \cdot g)' = f' \cdot g + g' \cdot f
$$

### Proof of the Product Rule

The proof of the Product Rule involves a strategic addition and subtraction within the derivative's limit definition. We manipulate the expression by adding and subtracting $f(x + h)g(x)$ in the numerator, allowing us to separate the limit into two parts that represent the derivatives of $f$ and $g$:

$$
\frac{d}{dx}[f(x)g(x)] = \lim_{h \to 0} \frac{f(x + h)g(x + h) - f(x)g(x)}{h}
$$

$$
= \lim_{h \to 0} \left( f(x + h) \frac{g(x + h) - g(x)}{h} + g(x) \frac{f(x + h) - f(x)}{h} \right)
$$

$$
= f(x) \lim_{h \to 0} \frac{g(x + h) - g(x)}{h} + g(x) \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

$$
= f(x) \frac{d}{dx}[g(x)] + g(x) \frac{d}{dx}[f(x)]
$$

In the final step, $f(x + h)$ approaches $f(x)$ as $h$ approaches $0$ because $f$ is continuous at $x$ by Theorem 2.2.3. Also, $g(x)$ is treated as a constant in the limit since it does not depend on $h$.

**In summary**, the derivative of a product of two functions is the sum of two terms: the first function times the derivative of the second, and the second function times the derivative of the first.


# Example Solutions for Derivatives

In these examples, we will find the derivative of given functions using different methods, including the product rule.

## Example 1: Find $\frac{dy}{dx}$ if $y = (4x^2 - 1)(7x^3 + x)$.

### Method 1: Using the Product Rule

Apply the product rule, which states that the derivative of a product $f(x)g(x)$ is $f'(x)g(x) + f(x)g'(x)$:

$$
\frac{dy}{dx} = \frac{d}{dx}[(4x^2 - 1)(7x^3 + x)] = (4x^2 - 1) \frac{d}{dx}[7x^3 + x] + (7x^3 + x) \frac{d}{dx}[4x^2 - 1]
$$

Calculating the derivatives inside the brackets:

$$
\frac{dy}{dx} = (4x^2 - 1)(21x^2 + 1) + (7x^3 + x)(8x)
$$

Simplify to get the final derivative:

$$
\frac{dy}{dx} = 140x^4 - 9x^2 - 1
$$

### Method 2: Multiplying First

Expand the expression for $y$ and then differentiate term by term:

$$
y = (4x^2 - 1)(7x^3 + x) = 28x^5 - 3x^3 - x
$$

Differentiate using the power rule:

$$
\frac{dy}{dx} = \frac{d}{dx}[28x^5 - 3x^3 - x] = 140x^4 - 9x^2 - 1
$$

Both methods yield the same result for $\frac{dy}{dx}$.

## Example 2: Find $\frac{ds}{dt}$ if $s = (1 + t)\sqrt{t}$.

### Applying the Product Rule

For the function $s(t) = (1 + t)\sqrt{t}$, apply the product rule:

$$
\frac{ds}{dt} = \frac{d}{dt}[(1 + t)\sqrt{t}] = (1 + t) \frac{d}{dt}[\sqrt{t}] + \sqrt{t} \frac{d}{dt}[1 + t]
$$

Calculate the derivatives:

$$
\frac{ds}{dt} = (1 + t) \frac{1}{2\sqrt{t}} + \sqrt{t}
$$

Simplify and combine like terms:

$$
\frac{ds}{dt} = \frac{1 + t}{2\sqrt{t}} + \sqrt{t} = \frac{1 + 3t}{2\sqrt{t}}
$$

The derivative $\frac{ds}{dt}$ is $\frac{1 + 3t}{2\sqrt{t}}$.


# Derivative of a Quotient

The derivative of a quotient of two functions is given by the Quotient Rule. Unlike a simple division of the derivatives, the Quotient Rule provides a way to differentiate a function that is the ratio of two other functions.

## The Quotient Rule

**Theorem 2.4.2 (The Quotient Rule):** If $f$ and $g$ are both differentiable at $x$ and $g(x) \neq 0$, then the quotient $f/g$ is differentiable at $x$, and its derivative is given by:

$$
\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right] = \frac{g(x)\frac{d}{dx}[f(x)] - f(x)\frac{d}{dx}[g(x)]}{[g(x)]^2}
$$

This can be symbolically expressed as:

$$
\left(\frac{f}{g}\right)' = \frac{g \cdot f' - f \cdot g'}{g^2}
$$

### Proof of the Quotient Rule

The proof involves adding and subtracting $f(x) \cdot g(x)$ within the limit process of the derivative definition for a quotient and then simplifying the result. The detailed steps are similar to those of the Product Rule, involving limits and continuity arguments.

## Example 3: Find $y'(x)$ for $y = \frac{x^3 + 2x^2 - 1}{x + 5}$.

**Solution:**

Apply the quotient rule to find the derivative of the function:

$$
\frac{dy}{dx} = \frac{(x + 5) \frac{d}{dx}[x^3 + 2x^2 - 1] - (x^3 + 2x^2 - 1) \frac{d}{dx}[x + 5]}{(x + 5)^2}
$$

Differentiate the numerator:

$$
\frac{dy}{dx} = \frac{(x + 5)(3x^2 + 4x) - (x^3 + 2x^2 - 1)(1)}{(x + 5)^2}
$$

Expand and simplify:

$$
\frac{dy}{dx} = \frac{(3x^3 + 19x^2 + 20x) - (x^3 + 2x^2 - 1)}{(x + 5)^2}
$$

Combine like terms:

$$
\frac{dy}{dx} = \frac{2x^3 + 17x^2 + 20x + 1}{(x + 5)^2}
$$

The derivative of the function $y = \frac{x^3 + 2x^2 - 1}{x + 5}$ is $\frac{2x^3 + 17x^2 + 20x + 1}{(x + 5)^2}$, which simplifies the rate of change of $y$ with respect to $x$.


# Example 4: Horizontal Tangent Lines of a Function

Let $f(x) = \frac{x^2 - 1}{x^4 + 1}$.

## Part (b): Finding Exact Locations of Horizontal Tangent Lines

To find the exact locations of the horizontal tangent lines, we find the points where $\frac{dy}{dx} = 0$ by differentiating $f(x)$.

**Solution:**

First, we apply the quotient rule for differentiation, which states that:

$$
\left(\frac{f}{g}\right)' = \frac{g \cdot f' - f \cdot g'}{g^2}
$$

For the given function $f(x)$, we have:

$$
\frac{dy}{dx} = \frac{(x^4 + 1)(2x) - (x^2 - 1)(4x^3)}{(x^4 + 1)^2}
$$

Simplify the expression:

$$
\frac{dy}{dx} = \frac{-2x^5 + 4x^3 + 2x}{(x^4 + 1)^2}
$$

Further simplification gives us:

$$
\frac{dy}{dx} = \frac{-2x(x^4 - 2x^2 - 1)}{(x^4 + 1)^2}
$$

Setting the derivative equal to zero:

$$
-2x(x^4 - 2x^2 - 1) = 0
$$

The solutions to this equation are the values of $x$ where the numerator is zero:

1. The first factor gives $x = 0$.
2. The other solutions are found by solving the quadratic in terms of $x^2$:
   $$
   x^4 - 2x^2 - 1 = 0
   $$
   Which can be solved to yield $x^2 = 1 \pm \sqrt{2}$. We disregard the negative root as it gives an imaginary result.

Thus, the exact solutions are:

$$
x = 0, x = \sqrt{1 + \sqrt{2}} \approx 1.55, \text{ and } x = -\sqrt{1 + \sqrt{2}} \approx -1.55
$$

These results align with our graphical estimates from part (a).

## Summary

In summary, horizontal tangent lines for the function $y = \frac{x^2 - 1}{x^4 + 1}$ occur at $x = 0$, $x \approx 1.55$, and $x \approx -1.55$.


# Summary of Differentiation Rules

The table below summarizes the differentiation rules for various types of functions.

| Rule | Derivative Formula | Notes |
| --- | ----- | ----- |
| Constant Rule | $\frac{d}{dx}[c] = 0$ | The derivative of a constant is zero. |
| Power Rule | $\frac{d}{dx}[x^r] = rx^{r-1}$ | Applies to functions of the form $x^r$. |
| Constant Multiple Rule | $(cf)' = cf'$ | The derivative of a constant times a function is the constant times the derivative of the function. |
| Sum Rule | $(f + g)' = f' + g'$ | The derivative of a sum is the sum of the derivatives. |
| Difference Rule | $(f - g)' = f' - g'$ | The derivative of a difference is the difference of the derivatives. |
| Product Rule | $(f \cdot g)' = f \cdot g' + g \cdot f'$ | The derivative of a product is not the product of the derivatives, but this combination. |
| Quotient Rule | $\left(\frac{f}{g}\right)' = \frac{g \cdot f' - f \cdot g'}{g^2}$ | The derivative of a quotient is given by this formula, assuming $g(x) \neq 0$. |
| Reciprocal Rule | $\left(\frac{1}{g}\right)' = -\frac{g'}{g^2}$ | Special case of the Quotient Rule when the numerator is 1. |

These rules form the foundation for differentiation in calculus and are used to find the rate of change or slope of various functions.


### Example 1: Product Rule

#### Problem Statement:
Find the derivative of $y = \frac{x^2 + x + 2}{x^3 + 6}$.

#### Solution:
Using the Quotient Rule, which states that if $y = \frac{u}{v}$, then $y' = \frac{vu' - uv'}{v^2}$, we get:

$$
y' = \frac{(x^3 + 6)(2x + 1) - (x^2 + x + 2)(3x^2)}{(x^3 + 6)^2}
$$

Expanding and simplifying:

$$
y' = \frac{2x^4 + x^3 + 12x + 6 - 3x^4 - 3x^3 - 6x^2}{(x^3 + 6)^2}
$$

Further simplifying:

$$
y' = \frac{-x^4 - 2x^3 + 6x^2 + 12x + 6}{(x^3 + 6)^2}
$$

### Example 2: Product Rule

#### Problem Statement:
Find $\frac{dy}{dx}$, where $y = f(x)$ and $f(x) = (x^3 + 7x^2 - 8)(2x^{-3} + x^{-4})$.

#### Solution:
Using the Product Rule, which states that if $y = uv$, then $y' = u'v + uv'$, we find $f'(x)$:

$$
f'(x) = \frac{d}{dx} [x^3 + 7x^2 - 8] \cdot (2x^{-3} + x^{-4}) + (x^3 + 7x^2 - 8) \cdot \frac{d}{dx} [2x^{-3} + x^{-4}]
$$

Expanding:

$$
f'(x) = (3x^2 + 14x) \cdot (2x^{-3} + x^{-4}) + (x^3 + 7x^2 - 8) \cdot (-6x^{-4} - 4x^{-5})
$$

Further simplification yields:

$$
f'(x) = -15x^{-2} - 14x^{-3} + 48x^{-4} + 32x^{-5}
$$

Therefore, $\frac{dy}{dx} = -15x^{-2} - 14x^{-3} + 48x^{-4} + 32x^{-5}$.


### Problem 3: Quotient Rule

#### Question:
Given $f(x) = \frac{3x + 4}{x^2 + 1}$, find $f'(x)$.

#### Answer:

Using the Quotient Rule, we have:

$$
f'(x) = \frac{(x^2 + 1)\frac{d}{dx}[3x + 4] - (3x + 4)\frac{d}{dx}[x^2 + 1]}{(x^2 + 1)^2}
$$

Simplifying further:

$$
f'(x) = \frac{(x^2 + 1)(3) - (3x + 4)(2x)}{(x^2 + 1)^2} = \frac{3x^2 + 3 - 6x^2 - 8x}{(x^2 + 1)^2}
$$

Finally, we get:

$$
f'(x) = \frac{-3x^2 - 8x + 3}{(x^2 + 1)^2}
$$

### Problem 4: Quotient Rule

#### Question:
Given $f(x) = \frac{x - 2}{x^4 + x + 1}$, find $f'(x)$.

#### Answer:

Applying the quotient rule, we have:

$$
f'(x) = \frac{(x^4 + x + 1)\frac{d}{dx}[x - 2] - (x - 2)\frac{d}{dx}[x^4 + x + 1]}{(x^4 + x + 1)^2}
$$

Simplifying further:

$$
f'(x) = \frac{(x^4 + x + 1)(1) - (x - 2)(4x^3 + 1)}{(x^4 + x + 1)^2} = \frac{x^4 + x + 1 - 4x^4 - 8x^3 - x + 2}{(x^4 + x + 1)^2}
$$

Finally, we get:

$$
f'(x) = \frac{-3x^4 + 8x^3 + 3}{(x^4 + x + 1)^2}
$$
