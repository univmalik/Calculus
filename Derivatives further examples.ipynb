### Example 1

Find a point on the curver $y = 2 - \sqrt{x}$ where the slope is $-0.5$. 



### Solution

Ee need to determine the $x$-coordinate where the derivative of the function equals $-0.5$. Once we find the $x$-coordinate, we can substitute it back into the original function to find the corresponding $y$-coordinate.

## Function and Derivative

Given the function:

$$
f(x) = 2 - \sqrt{x}
$$

To find the derivative, we first rewrite the function in a form suitable for applying the power rule:

$$
f(x) = 2 - x^{1/2}
$$

Using the power rule, the derivative of $x^{1/2}$ is:

$$
\frac{d}{dx} x^{1/2} = \frac{1}{2} x^{-1/2} = \frac{1}{2\sqrt{x}}
$$

Therefore, the derivative of $f(x)$ is:

$$
f'(x) = -\frac{1}{2} x^{-1/2} = -\frac{1}{2\sqrt{x}}
$$

## Finding the $x$-coordinate Where the Slope is $-0.5$

We set the derivative equal to $-0.5$ and solve for $x$:

$$
-\frac{1}{2\sqrt{x}} = -0.5
$$

First, remove the negative sign:

$$
\frac{1}{2\sqrt{x}} = 0.5
$$

Next, multiply both sides by $2\sqrt{x}$:

$$
1 = 2\sqrt{x} \cdot 0.5
$$

$$
1 = \sqrt{x}
$$

Square both sides to solve for $x$:

$$
1^2 = (\sqrt{x})^2
$$

$$
x = 1
$$

## Finding the Corresponding $y$-coordinate

Substitute $x = 1$ back into the original function to find $y$:

$$
y = 2 - \sqrt{1}
$$

$$
y = 2 - 1
$$

$$
y = 1
$$

## Conclusion

The point on the curve $y = 2 - \sqrt{x}$ where the slope is $-0.5$ is $(1, 1)$.

The line of the graph would then be $y - 1 = -0.5(x - 1)$

## Finding the $x$-coordinate Where the Slope is $0.4$

We need to determine the $x$-coordinate where the derivative of the function equals $0.4$. Once we find the $x$-coordinate, we can substitute it back into the original function to find the corresponding $y$-coordinate.

## Function and Derivative

Given the function:

$$
f(x) = \frac{x^3}{2} + 2
$$

To find the derivative, we use the power rule:

$$
f'(x) = \frac{d}{dx} \left( \frac{x^3}{2} \right) + \frac{d}{dx}(2)
$$

Since the derivative of a constant is zero, we get:

$$
f'(x) = \frac{3}{2} x^2
$$

## Finding the $x$-coordinate Where the Slope is $0.4$

We set the derivative equal to $0.4$ and solve for $x$:

$$
\frac{3}{2} x^2 = 0.4
$$

First, multiply both sides by $\frac{2}{3}$ to isolate $x^2$:

$$
x^2 = \frac{0.4 \times 2}{3} = \frac{0.8}{3} = \frac{4}{15}
$$

Next, take the square root of both sides to solve for $x$:

$$
x = \pm \sqrt{\frac{4}{15}} = \pm \frac{2}{\sqrt{15}} = \pm \frac{2 \sqrt{15}}{15}
$$

$$
x = \pm 0.516
$$

## Finding the Corresponding $y$-coordinate

Substitute $x = 0.516$ back into the original function to find $y$:

$$
y = \frac{(0.516)^3}{2} + 2
$$

Simplify the expression:

$$
y = \frac{0.137}{2} + 2
$$

$$
y = 0.0684 + 2
$$

$$
y = 2.07
$$

Similarly, substitute $x = -0.516$ back into the original function to find $y$:

$$
y = \frac{(-0.516)^3}{2} + 2
$$

Simplify the expression:

$$
y = \frac{-0.137}{2} + 2
$$

$$
y = -0.0684 + 2
$$

$$
y = 1.93
$$

## Conclusion

The points on the curve $y = \frac{x^3}{2} + 2$ where the slope is $0.4$ are approximately:

$$
\left( 0.516, 2.07 \right)
$$

and

$$
\left( -0.516, 1.93 \right)
$$
