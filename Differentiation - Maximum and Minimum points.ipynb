# Maximum and Minimum Points

## Introduction

In calculus, the maximum and minimum points of a function are critical points where the function reaches its highest or lowest values within a given interval. These points are also known as extrema (plural of extremum). 

To determine these points, we use the derivative of the function. A function $f(x)$ has a local maximum or minimum at a point $x = c$ if the derivative $f'(c) = 0$. This condition indicates that the tangent to the curve at this point is horizontal.

- The point on the right is a local minimum because the function changes from decreasing to increasing at this point. Again, the slope of the tangent line is zero, indicating a horizontal tangent.

## Finding Maximum and Minimum Points

To find the maximum and minimum points of a function $f(x)$:
1. Calculate the derivative $f'(x)$.
2. Solve the equation $f'(x) = 0$ to find the critical points.
3. Analyze the sign of $f'(x)$ around the critical points to determine whether each point is a maximum, minimum, or a saddle point (where the function changes concavity).




### Steps to Determine Maximum and Minimum Points

1. **Calculate the derivative of the function:**

   For example, if $f(x) = x^3 - 3x^2 + 4$:

   $$
   f'(x) = 3x^2 - 6x
   $$

2. **Set the derivative equal to zero and solve for $x$:**

   $$
   3x^2 - 6x = 0
   $$

   $$
   x(3x - 6) = 0
   $$

   $$
   x = 0 \quad \text{or} \quad x = 2
   $$

3. **Analyze the sign of $f'(x)$ around the critical points:**

   - For $x = 0$:
     - When $x$ is slightly less than 0, $f'(x)$ is positive.
     - When $x$ is slightly more than 0, $f'(x)$ is negative.
     - Therefore, $x = 0$ is a local maximum.

   - For $x = 2$:
     - When $x$ is slightly less than 2, $f'(x)$ is negative.
     - When $x$ is slightly more than 2, $f'(x)$ is positive.
     - Therefore, $x = 2$ is a local minimum.

4. **Conclusion:**

   The function $f(x) = x^3 - 3x^2 + 4$ has a local maximum at $x = 0$ and a local minimum at $x = 2$.

## Summary

Finding maximum and minimum points involves:
- Calculating the first derivative of the function.
- Setting the derivative equal to zero to find critical points.
- Analyzing the sign changes of the first derivative around these points to classify them as maxima or minima.


# Practice Questions on Maximum and Minimum Points

## Question 2

## Question 3

Find the maximum and minimum points of the function $f(x) = x^3 - 6x^2 + 9x + 1$.

### Solution 1:

1. **Calculate the derivative:**

   $$
   f'(x) = 3x^2 - 6x
   $$

2. **Set the derivative equal to zero and solve for $x$:**

   $$
   3x^2 - 6x = 0
   $$

   $$
   x(3x - 6) = 0
   $$

   $$
   x = 0 \quad \text{or} \quad x = 2
   $$

3. **Find the corresponding $y$-coordinates:**

   - For $x = 0$:
     $$
     f(0) = 0^3 - 3(0)^2 + 2 = 2
     $$

   - For $x = 2$:
     $$
     f(2) = 2^3 - 3(2)^2 + 2 = 8 - 12 + 2 = -2
     $$

4. **Conclusion:**

   The function $f(x) = x^3 - 3x^2 + 2$ has a local maximum at $(0, 2)$ and a local minimum at $(2, -2)$.

   

### Solution 2:

1. **Calculate the derivative:**

   $$
   f'(x) = 2x - 4
   $$

2. **Set the derivative equal to zero and solve for $x$:**

   $$
   2x - 4 = 0
   $$

   $$
   x = 2
   $$

3. **Find the corresponding $y$-coordinate:**

   $$
   f(2) = 2^2 - 4(2) + 4 = 4 - 8 + 4 = 0
   $$

4. **Conclusion:**

   The function $f(x) = x^2 - 4x + 4$ has a minimum point at $(2, 0)$.

   

### Solution 3:

1. **Calculate the derivative:**

   $$
   f'(x) = 3x^2 - 12x + 9
   $$

2. **Set the derivative equal to zero and solve for $x$:**

   $$
   3x^2 - 12x + 9 = 0
   $$

   $$
   x^2 - 4x + 3 = 0
   $$

   $$
   (x - 1)(x - 3) = 0
   $$

   $$
   x = 1 \quad \text{or} \quad x = 3
   $$

3. **Find the corresponding $y$-coordinates:**

   - For $x = 1$:
     $$
     f(1) = 1^3 - 6(1)^2 + 9(1) + 1 = 1 - 6 + 9 + 1 = 5
     $$

   - For $x = 3$:
     $$
     f(3) = 3^3 - 6(3)^2 + 9(3) + 1 = 27 - 54 + 27 + 1 = 1
     $$

4. **Conclusion:**

   The function $f(x) = x^3 - 6x^2 + 9x + 1$ has a local maximum at $(1, 5)$ and a local minimum at $(3, 1)$.
