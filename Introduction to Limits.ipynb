# Introduction to Limits in Mathematics

In calculus, the concept of a limit is fundamental. Limits help us understand the behavior of functions as they approach a specific point. They are essential in defining derivatives, integrals, and continuity.

## What is a Limit?

Intuitively, the limit of a function $f(x)$ as $x$ approaches a certain value $c$ is the value that $f(x)$ gets closer to as $x$ gets closer to $c$. We denote this as:

$$
\lim_{{x \to c}} f(x) = L
$$

This notation means that as $x$ approaches $c$, the function $f(x)$ approaches $L$.

### Example of a Limit

Consider the function $f(x) = 2x$. To find the limit as $x$ approaches 3, we write:

$$
\lim_{{x \to 3}} 2x
$$

As $x$ gets closer to 3, $2x$ gets closer to $6$. Therefore,

$$
\lim_{{x \to 3}} 2x = 6
$$

## Left-Hand and Right-Hand Limits

Limits can also be approached from one side: either from the left or from the right. These are called the left-hand limit and the right-hand limit, respectively.

- The left-hand limit of $f(x)$ as $x$ approaches $c$ is denoted as:

$$
\lim_{{x \to c^-}} f(x)
$$

- The right-hand limit of $f(x)$ as $x$ approaches $c$ is denoted as:

$$
\lim_{{x \to c^+}} f(x)
$$

For the limit to exist at a point, both the left-hand and right-hand limits must exist and be equal.

### Example of One-Sided Limits

Consider the piecewise function:

$$
f(x) = 
\begin{cases} 
x^2 & \text{if } x < 1 \\
2x & \text{if } x \geq 1 
\end{cases}
$$



To find the left-hand limit as $x$ approaches 1:

$$
\lim_{{x \to 1^-}} f(x) = \lim_{{x \to 1^-}} x^2 = 1^2 = 1
$$

To find the right-hand limit as $x$ approaches 1:

$$
\lim_{{x \to 1^+}} f(x) = \lim_{{x \to 1^+}} 2x = 2 \cdot 1 = 2
$$

Since the left-hand limit ($1$) is not equal to the right-hand limit ($2$), the limit of $f(x)$ as $x$ approaches 1 does not exist.

## Infinite Limits

Sometimes, as $x$ approaches a certain value, the function $f(x)$ increases or decreases without bound. This is called an infinite limit. For example:

$$
\lim_{{x \to 0}} \frac{1}{x^2} = \infty
$$

As $x$ approaches $0$, $\frac{1}{x^2}$ grows larger and larger without bound, hence the limit is infinity.








# Further Examples of Limits Approaching Negative Infinity

In addition to positive infinity, limits can also approach negative infinity. This occurs when a function decreases without bound as the input approaches a certain value. Here are some examples of such limits.

## Example 1: Rational Function

Consider the function:

$$
f(x) = \frac{-1}{x}
$$



To find the limit as $x$ approaches $0$ from the positive side and the negative side, we write:

### As $x$ approaches $0$ from the positive side:

$$
\lim_{{x \to 0^+}} \frac{-1}{x}
$$

As $x$ approaches $0$ from the right (positive side), $\frac{-1}{x}$ decreases without bound. Thus,

$$
\lim_{{x \to 0^+}} \frac{-1}{x} = -\infty
$$

### As $x$ approaches $0$ from the negative side:

$$
\lim_{{x \to 0^-}} \frac{-1}{x}
$$

As $x$ approaches $0$ from the left (negative side), $\frac{-1}{x}$ increases without bound. Thus,

$$
\lim_{{x \to 0^-}} \frac{-1}{x} = \infty
$$

This illustrates that the limit approaching zero from different sides can yield different results, including negative infinity.

## Example 2: Exponential Function

Consider the function:

$$
f(x) = -e^x
$$



To find the limit as $x$ approaches positive infinity, we write:

$$
\lim_{{x \to \infty}} -e^x
$$

As $x$ grows larger and larger, $e^x$ increases exponentially. Multiplying by $-1$ makes the function decrease without bound. Thus,

$$
\lim_{{x \to \infty}} -e^x = -\infty
$$

## Example 3: Polynomial Function

Consider the function:

$$
f(x) = -x^3
$$



To find the limit as $x$ approaches positive infinity, we write:

$$
\lim_{{x \to \infty}} -x^3
$$

As $x$ grows larger and larger, $x^3$ increases very rapidly. Multiplying by $-1$ causes the function to decrease without bound. Thus,

$$
\lim_{{x \to \infty}} -x^3 = -\infty
$$

Similarly, to find the limit as $x$ approaches negative infinity, we write:

$$
\lim_{{x \to -\infty}} -x^3
$$

As $x$ becomes more negative, $x^3$ becomes more negative as well. Multiplying by $-1$ makes the function increase without bound. Thus,

$$
\lim_{{x \to -\infty}} -x^3 = \infty
$$




# Example of a Limit Approaching 0

In addition to limits approaching infinity, limits can also approach a finite value such as 0. Let's consider an example of a limit reaching 0.

## Example: Rational Function

Consider the function:

$$
f(x) = \frac{1}{x^2}
$$



To find the limit as $x$ approaches infinity, we write:

$$
\lim_{{x \to \infty}} \frac{1}{x^2}$$

As $x$ grows larger and larger, the denominator $x^2$ increases very rapidly. Since the numerator is constant (1), the fraction $\frac{1}{x^2}$ gets smaller and smaller, approaching 0. Thus,
$$
\lim_{{x \to \infty}} \frac{1}{x^2} = 0$$

This example illustrates how a function can approach 0 as the input becomes very large.
