# 13.6 Directional Derivatives and Gradients

Directional derivatives extend the concept of partial derivatives to the rate of change of a function in any direction. Unlike partial derivatives which are limited to the coordinate axes, directional derivatives can describe how a function changes in any given direction.

## Directional Derivatives

The rate of change of a function $f(x, y)$ from a point $(x_0, y_0)$ in any direction can be described using a unit vector $\vec{u}$.

- A unit vector in the direction of interest: $\vec{u} = u_1\vec{i} + u_2\vec{j}$
- The line $l$ in the $xy$-plane is given by $x = x_0 + su_1, y = y_0 + su_2$, where $s$ is the arc length parameter from $(x_0, y_0)$.



_Figure 13.6.1_ illustrates the unit vector $\vec{u}$ that defines a specific direction from the point $(x_0, y_0)$.

On the line $l$, $z$ as a function of $s$ is expressed by $z = f(x_0 + su_1, y_0 + su_2)$. The derivative $\frac{dz}{ds}$ at $s = 0$ gives the rate of change of $f(x, y)$ with respect to distance in the direction of $\vec{u}$.

### Definition of Directional Derivative

If $f(x, y)$ is a function of $x$ and $y$, and $\vec{u} = u_1\vec{i} + u_2\vec{j}$ is a unit vector, the directional derivative of $f$ in the direction of $\vec{u}$ at $(x_0, y_0)$ is denoted by $D_{\vec{u}}f(x_0, y_0)$ and is defined by:

$$
D_{\vec{u}}f(x_0, y_0) = \left. \frac{d}{ds} f(x_0 + su_1, y_0 + su_2) \right|_{s=0}
$$

provided this derivative exists.



_Figure 13.6.2_ shows the slope of the surface $z = f(x, y)$ in the direction of $\vec{u}$ at the point $(x_0, y_0, f(x_0, y_0))$. This slope is the directional derivative and indicates how the function $f$ changes in the direction of $\vec{u}$ from the point $(x_0, y_0)$.

The value of $D_{\vec{u}}f(x_0, y_0)$ depends on both the location $(x_0, y_0)$ and the direction $\vec{u}$. Thus, at a fixed point, the slope of the surface may vary depending on the direction.


# Example of Directional Derivatives

Given a function $f(x, y) = xy$, we want to find and interpret the directional derivative $D_{\vec{u}}f(1, 2)$ for the unit vector $\vec{u} = \frac{\sqrt{3}}{2}\vec{i} + \frac{1}{2}\vec{j}$.

## Solution

According to the definition, the directional derivative is calculated as follows:

$$
D_{\vec{u}}f(1, 2) = \left. \frac{d}{ds} f\left(1 + \frac{\sqrt{3}s}{2}, 2 + \frac{s}{2}\right) \right|_{s=0}
$$

Evaluating the function $f(x, y)$ at the point along the line in the direction of $\vec{u}$:

$$
f\left(1 + \frac{\sqrt{3}s}{2}, 2 + \frac{s}{2}\right) = \left(1 + \frac{\sqrt{3}s}{2}\right)\left(2 + \frac{s}{2}\right) = \frac{\sqrt{3}}{4}s^2 + \left(\frac{1}{2} + \sqrt{3}\right)s + 2
$$

Taking the derivative with respect to $s$ and evaluating at $s=0$ gives:

$$
D_{\vec{u}}f(1, 2) = \left. \left(\frac{\sqrt{3}}{2}s + \frac{1}{2} + \sqrt{3}\right) \right|_{s=0} = \frac{1}{2} + \sqrt{3}
$$

Therefore:

$$
D_{\vec{u}}f(1, 2) \approx 2.23
$$

This result implies that if we move a small distance from the point $(1, 2)$ in the direction of $\vec{u}$, the function $f(x, y) = xy$ will increase approximately by $2.23$ times the distance moved.



_Figure 13.6.3_ depicts that the slope of the surface $z = f(x, y)$ varies with the direction of $\vec{u}$. The value of the directional derivative $D_{\vec{u}}f(1, 2)$ indicates the rate at which $f(x, y)$ increases in the direction of $\vec{u}$.


# Directional Derivatives for Functions of Three Variables

For a function $f(x, y, z)$ of three variables, the concept of directional derivatives is extended similarly to functions of two variables.

## Definition 13.6.2 - Directional Derivative in Three Dimensions

If $\vec{u} = u_1\vec{i} + u_2\vec{j} + u_3\vec{k}$ is a unit vector and $f(x, y, z)$ is a function of $x$, $y$, and $z$, then the directional derivative of $f$ in the direction of $\vec{u}$ at $(x_0, y_0, z_0)$ is denoted by $D_{\vec{u}}f(x_0, y_0, z_0)$ and is defined as:

$$
D_{\vec{u}}f(x_0, y_0, z_0) = \left. \frac{d}{ds} f(x_0 + su_1, y_0 + su_2, z_0 + su_3) \right|_{s=0}
$$

provided this derivative exists.

While this equation does not lend itself to a simple geometric interpretation, it still represents the instantaneous rate of change of the function $f$ in a specified direction for functions of three variables.

## Theorem 13.6.3 - Computing Directional Derivatives

(a) If $f(x, y)$ is differentiable at $(x_0, y_0)$, and $\vec{u} = u_1\vec{i} + u_2\vec{j}$ is a unit vector, then the directional derivative $D_{\vec{u}}f(x_0, y_0)$ exists and is given by:

$$
D_{\vec{u}}f(x_0, y_0) = f_x(x_0, y_0)u_1 + f_y(x_0, y_0)u_2
$$

(b) If $f(x, y, z)$ is differentiable at $(x_0, y_0, z_0)$, and $\vec{u} = u_1\vec{i} + u_2\vec{j} + u_3\vec{k}$ is a unit vector, then the directional derivative $D_{\vec{u}}f(x_0, y_0, z_0)$ exists and is given by:

$$
D_{\vec{u}}f(x_0, y_0, z_0) = f_x(x_0, y_0, z_0)u_1 + f_y(x_0, y_0, z_0)u_2 + f_z(x_0, y_0, z_0)u_3
$$

### Proof for Part (a)

The proof utilizes the chain rule for the composition of functions and confirms that the directional derivative can be computed using first-order partial derivatives.

This theorem can be used to confirm the result of Example 1 for $f(x, y) = xy$. For this function, we have $f_x(1, 2) = 2$ and $f_y(1, 2) = 1$. With $\vec{u} = \frac{\sqrt{3}}{2}\vec{i} + \frac{1}{2}\vec{j}$, we apply the formula from part (a) to obtain the same result as in Example 1:

$$
D_{\vec{u}}f(1, 2) = 2\left(\frac{\sqrt{3}}{2}\right) + 1\left(\frac{1}{2}\right) = \frac{\sqrt{3} + 1}{2}
$$

### Angle Representation

A unit vector $\vec{u}$ in the $xy$-plane can be expressed in terms of an angle $\varphi$ as $\vec{u} = \cos(\varphi)\vec{i} + \sin(\varphi)\vec{j}$. Thus, the directional derivative can also be written as:

$$
D_{\vec{u}}f(x_0, y_0) = f_x(x_0, y_0)\cos(\varphi) + f_y(x_0, y_0)\sin(\varphi)
$$

This representation is particularly useful when the direction is given as an angle from the positive $x$-axis.
