# Power Rule in Differentiation

## Introduction

The power rule is one of the most commonly used rules in differentiation. It provides a straightforward way to find the derivative of functions of the form $f(x) = x^n$, where $n$ is any real number.

## Power Rule

The power rule states that if $f(x) = x^n$, where $n$ is a real number, then the derivative of $f(x)$ with respect to $x$ is:

$$
f'(x) = nx^{n-1}
$$

This rule simplifies the process of finding derivatives of polynomial functions and functions with fractional or negative exponents.

## Easy Examples

### Example 1: $f(x) = x^2$

Apply the power rule with $n = 2$:

$$
f(x) = x^2
$$

Using the power rule:

$$
f'(x) = 2x^{2-1} = 2x
$$

### Example 2: $f(x) = 2x^2$

Apply the power rule with $n = 2$ and multiply by the constant coefficient 2:

$$
f(x) = 2x^2
$$

Using the power rule:

$$
f'(x) = 2 \cdot 2x^{2-1} = 4x
$$

## Intermediate Examples

### Example 3: $f(x) = x^{1/2}$

Apply the power rule with $n = 1/2$:

$$
f(x) = x^{1/2}
$$

Using the power rule:

$$
f'(x) = \frac{1}{2} x^{1/2 - 1} = \frac{1}{2} x^{-1/2}
$$

### Example 4: $f(x) = x^{-1/2}$

Apply the power rule with $n = -1/2$:

$$
f(x) = x^{-1/2}
$$

Using the power rule:

$$
f'(x) = -\frac{1}{2} x^{-1/2 - 1} = -\frac{1}{2} x^{-3/2}
$$

## Advanced Examples

### Example 5: $f(x) = 4x^4$

Apply the power rule with $n = 4$ and multiply by the constant coefficient 4:

$$
f(x) = 4x^4
$$

Using the power rule:

$$
f'(x) = 4 \cdot 4x^{4-1} = 16x^3
$$

### Example 6: $f(x) = 3x^{-1/3}$

Apply the power rule with $n = -1/3$ and multiply by the constant coefficient 3:

$$
f(x) = 3x^{-1/3}
$$

Using the power rule:

$$
f'(x) = 3 \cdot \left(-\frac{1}{3}\right)x^{-1/3 - 1} = -x^{-4/3}
$$

## Conclusion

The power rule is a fundamental tool in differentiation that allows us to easily find the derivatives of functions of the form $x^n$. Here are the key steps to apply the power rule:
1. Identify the exponent $n$ in the function $x^n$.
2. Multiply by the exponent $n$.
3. Subtract 1 from the exponent to find the new exponent.

Using these steps, we can quickly and efficiently find the derivatives of a wide variety of functions. Here are some examples:

- For $f(x) = x^2$, the derivative is $f'(x) = 2x$.
- For $f(x) = 2x^2$, the derivative is $f'(x) = 4x$.
- For $f(x) = x^{1/2}$, the derivative is $f'(x) = \frac{1}{2} x^{-1/2}$.
- For $f(x) = x^{-1/2}$, the derivative is $f'(x) = -\frac{1}{2} x^{-3/2}$.
- For $f(x) = 4x^4$, the derivative is $f'(x) = 16x^3$.
- For $f(x) = 3x^{-1/3}$, the derivative is $f'(x) = -x^{-4/3}$.

Understanding and applying the power rule is essential for solving many problems in calculus and related fields.

# Finding Equations of Tangent Lines to Curves at Particular Points

## Introduction

The equation of a tangent line to a curve at a particular point can be found using the derivative of the function. The tangent line at a point $(a, f(a))$ has the same slope as the derivative of the function at that point. The general form of the tangent line equation is:

$$y - y_1 = m(x - x_1)$$

$$
y - f(a) = f'(a)(x - a)
$$

## Example 1: $y = x^{1/2}$ at $x = 2$

### Step 1: Find the derivative of the function

Given the function:

$$
f(x) = x^{1/2}
$$

We find the derivative using the power rule:

$$
f'(x) = \frac{1}{2} x^{1/2 - 1} = \frac{1}{2} x^{-1/2}
$$

### Step 2: Evaluate the derivative at $x = 2$

To find the slope of the tangent line at $x = 2$:

$$
f'(2) = \frac{1}{2} \cdot 2^{-1/2} = \frac{1}{2\sqrt{2}} = \frac{1}{2\sqrt{2}} = \frac{1}{2\sqrt{2}} \cdot \frac{\sqrt{2}}{\sqrt{2}} = \frac{\sqrt{2}}{4}
$$

### Step 3: Find the value of the function at $x = 2$

To find the point on the curve:

$$
f(2) = 2^{1/2} = \sqrt{2}
$$

The point of tangency is $(2, \sqrt{2})$.

### Step 4: Write the equation of the tangent line

Using the point-slope form of the line:

$$
y - f(a) = f'(a)(x - a)
$$

Substitute $a = 2$, $f(a) = \sqrt{2}$, and $f'(a) = \frac{\sqrt{2}}{4}$:

$$
y - \sqrt{2} = \frac{\sqrt{2}}{4}(x - 2)
$$

Simplify:

$$
y = \frac{\sqrt{2}}{4}x - \frac{\sqrt{2}}{4} \cdot 2 + \sqrt{2}
$$

$$
y = \frac{\sqrt{2}}{4}x - \frac{\sqrt{2}}{2} + \sqrt{2}
$$

$$
y = \frac{\sqrt{2}}{4}x + \frac{\sqrt{2}}{2}
$$

So, the equation of the tangent line to the curve $y = x^{1/2}$ at $x = 2$ is:

$$
y = \frac{\sqrt{2}}{4}x + \frac{\sqrt{2}}{2}
$$



## Example 2: $y = 4x^3$ at $x = 0.5$

### Step 1: Find the derivative of the function

Given the function:

$$
f(x) = 4x^3
$$

We find the derivative using the power rule:

$$
f'(x) = 4 \cdot 3x^{3-1} = 12x^2
$$

### Step 2: Evaluate the derivative at $x = 0.5$

To find the slope of the tangent line at $x = 0.5$:

$$
f'(0.5) = 12 \cdot (0.5)^2 = 12 \cdot 0.25 = 3
$$

### Step 3: Find the value of the function at $x = 0.5$

To find the point on the curve:

$$
f(0.5) = 4 \cdot (0.5)^3 = 4 \cdot 0.125 = 0.5
$$

The point of tangency is $(0.5, 0.5)$.

### Step 4: Write the equation of the tangent line

Using the point-slope form of the line:

$$
y - f(a) = f'(a)(x - a)
$$

Substitute $a = 0.5$, $f(a) = 0.5$, and $f'(a) = 3$:

$$
y - 0.5 = 3(x - 0.5)
$$

Simplify:

$$
y - 0.5 = 3x - 1.5
$$

$$
y = 3x - 1
$$

So, the equation of the tangent line to the curve $y = 4x^3$ at $x = 0.5$ is:

$$
y = 3x - 1
$$


## Conclusion

To find the equation of a tangent line to a curve at a particular point, we:
1. Find the derivative of the function.
2. Evaluate the derivative at the given point to find the slope.
3. Evaluate the function at the given point to find the coordinates of the point of tangency.
4. Use the point-slope form of the line to write the equation of the tangent line.

These steps help us determine the equation of the tangent line that touches the curve at a specific point.

## Solutions

### Solution 1: $ y = 3x^{-3} $ at $ x = 2 $

1. **Find the derivative**:

   $$
   f(x) = 3x^{-3}
   $$

   Using the power rule:

   $$
   f'(x) = 3 \cdot (-3)x^{-4} = -9x^{-4}
   $$

2. **Evaluate the derivative at $ x = 2 $**:

   $$
   f'(2) = -9(2)^{-4} = -9 \cdot \frac{1}{16} = -\frac{9}{16}
   $$

3. **Find the value of the function at $ x = 2 $**:

   $$
   f(2) = 3(2)^{-3} = 3 \cdot \frac{1}{8} = \frac{3}{8}
   $$

   The point of tangency is $ (2, \frac{3}{8}) $.

4. **Write the equation of the tangent line**:

   Using the point-slope form:

   $$
   y - \frac{3}{8} = -\frac{9}{16}(x - 2)
   $$

### Solution 2: $ y = 2x^2 - 2 $ at $ x = 1 $

1. **Find the derivative**:

   $$
   f(x) = 2x^2 - 2
   $$

   Using the power rule:

   $$
   f'(x) = 4x
   $$

2. **Evaluate the derivative at $ x = 1 $**:

   $$
   f'(1) = 4 \cdot 1 = 4
   $$

3. **Find the value of the function at $ x = 1 $**:

   $$
   f(1) = 2(1)^2 - 2 = 0
   $$

   The point of tangency is $ (1, 0) $.

4. **Write the equation of the tangent line**:

   Using the point-slope form:

   $$
   y - 0 = 4(x - 1)
   $$

   Simplify:

   $$
   y = 4x - 4
   $$

### Solution 3: $ y = \frac{4}{x^{1/2}} $ at $ x = 2 $

1. **Find the derivative**:

   $$
   f(x) = 4x^{-1/2}
   $$

   Using the power rule:

   $$
   f'(x) = 4 \cdot -\frac{1}{2}x^{-3/2} = -2x^{-3/2}
   $$

2. **Evaluate the derivative at $ x = 2 $**:

   $$
   f'(2) = -2(2)^{-3/2} = -2 \cdot \frac{1}{2\sqrt{2}} = -\frac{1}{2\sqrt{2}} = -\frac{1}{2\sqrt{2}} \cdot \frac{\sqrt{2}}{\sqrt{2}} = -\frac{\sqrt{2}}{4}
   $$

3. **Find the value of the function at $ x = 2 $**:

   $$
   f(2) = \frac{4}{2^{1/2}} = \frac{4}{\sqrt{2}} = 4 \cdot \frac{\sqrt{2}}{2} = 2\sqrt{2}
   $$

   The point of tangency is $ (2, 2\sqrt{2}) $.

4. **Write the equation of the tangent line**:

   Using the point-slope form:

   $$
   y - 2\sqrt{2} = -\frac{\sqrt{2}}{4}(x - 2)
   $$

### Solution 4: $ y = -3x^3 + 4 $ at $ x = 1 $

1. **Find the derivative**:

   $$
   f(x) = -3x^3 + 4
   $$

   Using the power rule:

   $$
   f'(x) = -9x^2
   $$

2. **Evaluate the derivative at $ x = 1 $**:

   $$
   f'(1) = -9(1)^2 = -9
   $$

3. **Find the value of the function at $ x = 1 $**:

   $$
   f(1) = -3(1)^3 + 4 = 1
   $$

   The point of tangency is $ (1, 1) $.

4. **Write the equation of the tangent line**:

   Using the point-slope form:

   $$
   y - 1 = -9(x - 1)
   $$

### Solution 5: $ y = 5x^{2.5} $ at $ x = 1 $

1. **Find the derivative**:

   $$
   f(x) = 5x^{2.5}
   $$

   Using the power rule:

   $$
   f'(x) = 5 \cdot 2.5x^{1.5} = 12.5x^{1.5}
   $$

2. **Evaluate the derivative at $ x = 1 $**:

   $$
   f'(1) = 12.5(1)^{1.5} = 12.5
   $$

3. **Find the value of the function at $ x = 1 $**:

   $$
   f(1) = 5(1)^{2.5} = 5
   $$

   The point of tangency is $ (1, 5) $.

4. **Write the equation of the tangent line**:

   Using the point-slope form:

   $$
   y - 5 = 12.5(x - 1)
   $$
