# Basic Integration

## Differentiation Reversed

When $x^2$ is differentiated with respect to $x$, the derivative is $2x$.

Conversely, if the derivative of an unknown function is $2x$, then the unknown function could be $x^2$.

This process of finding a function from its derivative, which reverses the operation of differentiating, is called **integration**.

## The Constant of Integration

As seen above, $2x$ is the derivative of $x^2$, but it is also the derivative of $x^2 + 3$, $x^2 - 9$, and, in fact, the derivative of $x^2$ plus any constant.

Therefore, the result of integrating $2x$, which is called the **integral** of $2x$, is not a unique function but is of the form:

$$ x^2 + K $$

where $K$ is any constant. 

$K$ is called the **constant of integration**.

This is written as:

$$ \int 2x \, dx = x^2 + K $$

where $\int \ldots \, dx$ means the integral of $\ldots$ with respect to $x$.

Integrating any function reverses the process of differentiating, so for any function $f(x)$ we have:

$$ \int \frac{d}{dx} f(x) \, dx = f(x) + K $$

For example, because differentiating $x^3$ with respect to $x$ gives $3x^2$, we have:

$$ \int 3x^2 \, dx = x^3 + K $$

and it follows that:

$$ \int x^2 \, dx = \frac{1}{3}x^3 + K $$

Note that it is not necessary to write $\frac{1}{3}K$ in the second form, as $K$ represents any constant in either expression.

In general, the derivative of $x^{n+1}$ is $(n+1)x^n$, so:

$$ \int x^n \, dx = \frac{1}{n+1} x^{n+1} + K $$

To integrate a power of $x$, increase the power by $1$ and divide by the new power.

This rule can be used to integrate any power of $x$ except $-1$, which is considered later.

### Integrating a Function with a Constant Multiplier

When integrating a function of the form $c x^n$, where $c$ is a constant, the constant can be factored out of the integral. This is because the integral of a constant multiplied by a function is equal to the constant multiplied by the integral of the function.

#### Example: Integrate $c x^n$
1. **Integrate $c x^n$**:
    $$ \int c x^n \, dx = c \int x^n \, dx $$

Using the rule for integrating $x^n$:
    $$ \int x^n \, dx = \frac{1}{n+1} x^{n+1} + K $$

So:
    $$ \int c x^n \, dx = c \left( \frac{1}{n+1} x^{n+1} \right) + K = \frac{c}{n+1} x^{n+1} + K $$

By following these rules, you can integrate polynomials and other functions of $x$ with a constant multiplier.

## Examples

1. **Integrate $x^3$**:
    $$ \int x^3 \, dx = \frac{1}{3+1} x^{3+1} + K = \frac{1}{4} x^4 + K $$

2. **Integrate $5x^4$**:
    $$ \int 5x^4 \, dx = 5 \int x^4 \, dx = 5 \left( \frac{1}{4+1} x^{4+1} \right) + K = \frac{5}{5} x^5 + K = x^5 + K $$

3. **Integrate $7x^{-2}$**:
    $$ \int 7x^{-2} \, dx = 7 \int x^{-2} \, dx = 7 \left( \frac{1}{-2+1} x^{-2+1} \right) + K = 7 \left( \frac{1}{-1} x^{-1} \right) + K = -7 x^{-1} + K = -\frac{7}{x} + K $$

4. **Integrate $4x^{1/2}$**:
    $$ \int 4x^{1/2} \, dx = 4 \int x^{1/2} \, dx = 4 \left( \frac{1}{1/2+1} x^{1/2+1} \right) + K = 4 \left( \frac{1}{3/2} x^{3/2} \right) + K = 4 \left( \frac{2}{3} x^{3/2} \right) + K = \frac{8}{3} x^{3/2} + K $$

# Integrating a Sum or Difference of Functions

We saw in Chapter 22 that a function can be differentiated term by term. Therefore, as integration reverses differentiation, integration also can be done term by term. This is known as the **addition rule** in integration.

## Addition Rule in Integration

If $f(x)$ and $g(x)$ are two functions, then the integral of their sum (or difference) is the sum (or difference) of their integrals. Mathematically, this is expressed as:

$$ \int (f(x) + g(x)) \, dx = \int f(x) \, dx + \int g(x) \, dx $$

and

$$ \int (f(x) - g(x)) \, dx = \int f(x) \, dx - \int g(x) \, dx $$

This rule allows us to break down more complex integrals into simpler parts that can be integrated individually.

## Example

### Find the integral of $x^7 + \frac{1}{x^2} - \sqrt{x}$

First, rewrite the integral:

$$ \int \left( x^7 + \frac{1}{x^2} - \sqrt{x} \right) \, dx = \int \left( x^7 + x^{-2} - x^{1/2} \right) \, dx $$

Now, integrate each term separately:

$$ \int x^7 \, dx + \int x^{-2} \, dx - \int x^{1/2} \, dx $$

Using the power rule for integration:

- For $\int x^7 \, dx$:
    $$ \int x^7 \, dx = \frac{1}{7+1} x^{7+1} + K = \frac{1}{8} x^8 + K $$

- For $\int x^{-2} \, dx$:
    $$ \int x^{-2} \, dx = \frac{1}{-2+1} x^{-2+1} + K = \frac{1}{-1} x^{-1} + K = -x^{-1} + K = -\frac{1}{x} + K $$

- For $\int x^{1/2} \, dx$:
    $$ \int x^{1/2} \, dx = \frac{1}{1/2+1} x^{1/2+1} + K = \frac{1}{3/2} x^{3/2} + K = \frac{2}{3} x^{3/2} + K $$

Combining these results:

$$ \int \left( x^7 + \frac{1}{x^2} - \sqrt{x} \right) \, dx = \frac{1}{8} x^8 - \frac{1}{x} - \frac{2}{3} x^{3/2} + K $$

By applying the addition rule, we can integrate sums or differences of functions term by term and then combine the results.


## Additional Integrals

### 5. Integrate $\frac{1}{x^{5/2}}$
$$ \int \frac{1}{x^{5/2}} \, dx = \int x^{-5/2} \, dx $$
$$ \int x^{-5/2} \, dx = \frac{1}{-5/2 + 1} x^{-5/2 + 1} + K = \frac{1}{-3/2} x^{-3/2} + K = -\frac{2}{3} x^{-3/2} + K $$

### 6. Integrate $x^{-1/2}$
$$ \int x^{-1/2} \, dx = \int x^{-1/2} \, dx $$
$$ \int x^{-1/2} \, dx = \frac{1}{-1/2 + 1} x^{-1/2 + 1} + K = \frac{1}{1/2} x^{1/2} + K = 2 x^{1/2} + K = 2 \sqrt{x} + K $$

### 7. Integrate $x$
$$ \int x \, dx = \int x^1 \, dx $$
$$ \int x \, dx = \frac{1}{1+1} x^{1+1} + K = \frac{1}{2} x^2 + K $$

### 8. Integrate $\frac{1}{\sqrt[3]{x}}$
$$ \int \frac{1}{\sqrt[3]{x}} \, dx = \int x^{-1/3} \, dx $$
$$ \int x^{-1/3} \, dx = \frac{1}{-1/3 + 1} x^{-1/3 + 1} + K = \frac{1}{2/3} x^{2/3} + K = \frac{3}{2} x^{2/3} + K $$

By following these steps, you can integrate sums or differences of functions term by term and apply the power rule for various types of integrals.
