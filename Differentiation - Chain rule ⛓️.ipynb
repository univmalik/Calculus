# Differentiation: Chain Rule ⛓️

## Introduction

The chain rule is a fundamental technique in calculus used to differentiate composite functions. It states that if a function $y(x)$ depends on $u(x)$, and $u(x)$ depends on $x$, then the derivative of $y(x)$ with respect to $x$ can be found by multiplying the derivative of $y(x)$ with respect to $u$ by the derivative of $u(x)$ with respect to $x$.

The chain rule (without proof here) can be expressed as:

$y = u(x)$,

$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx}
$$

This rule is particularly useful for differentiating functions that are compositions of other functions.

## Examples

### Example 1: Find $\frac{dy}{dx}$ of $y = (x^2 + 1)^{3/2}$

Let's differentiate the function $y = (x^2 + 1)^{3/2}$. 

We will let $y = u(x)$, then $u(x) = x^2 + 1$, then $y = u^{3/2}$.

1. **Identify the inner function and the outer function:**
   - Inner function: $u = x^2 + 1$
   - Outer function: $y = u^{3/2}$

2. **Differentiate the inner function with respect to $x$:**

$$
\frac{du}{dx} = \frac{d}{dx} (x^2 + 1) = 2x
$$

3. **Differentiate the outer function with respect to $u$:**

$$
\frac{dy}{du} = \frac{d}{du} (u^{3/2}) = \frac{3}{2} u^{1/2}
$$

4. **Apply the chain rule:**

$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} = \frac{3}{2} u^{1/2} \cdot 2x
$$

5. **Substitute back $u = x^2 + 1$:**

$$
\frac{dy}{dx} = \frac{3}{2} (x^2 + 1)^{1/2} \cdot 2x = 3x (x^2 + 1)^{1/2}
$$

### Example 2: $y = \frac{1}{(x^3 + 4)^2}$

Let's differentiate the function $y = \frac{1}{(x^3 + 4)^2}$.

1. **Rewrite the function in a more convenient form:**

$$
y = (x^3 + 4)^{-2}
$$

2. **Identify the inner function and the outer function:**
   - Inner function: $u = x^3 + 4$
   - Outer function: $y = u^{-2}$

3. **Differentiate the inner function with respect to $x$:**

$$
\frac{du}{dx} = \frac{d}{dx} (x^3 + 4) = 3x^2
$$

4. **Differentiate the outer function with respect to $u$:**

$$
\frac{dy}{du} = \frac{d}{du} (u^{-2}) = -2u^{-3}
$$

5. **Apply the chain rule:**

$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} = -2u^{-3} \cdot 3x^2
$$

6. **Substitute back $u = x^3 + 4$:**

$$
\frac{dy}{dx} = -2(x^3 + 4)^{-3} \cdot 3x^2 = -6x^2 (x^3 + 4)^{-3}
$$

## Conclusion

The chain rule is an essential tool for differentiating composite functions. By breaking down a complex function into its inner and outer functions, we can apply the chain rule to find the derivative. This technique simplifies the differentiation process and allows us to handle a wide variety of functions.

$$
\frac{du}{dx} = \frac{d}{dx} (2x^3 + 5) = 6x^2

$$

## Derivatives of $\sin(x)$ and $\cos(x)$

The derivative of $\sin(x)$ is $\cos(x)$:

$$
\frac{d}{dx} \sin(x) = \cos(x)
$$

The derivative of $\cos(x)$ is $-\sin(x)$:

$$
\frac{d}{dx} \cos(x) = -\sin(x)
$$

## Examples

### Example 1: $\frac{d}{dx} \sin(x^2 + x + 1)$

Let's differentiate the function $y = \sin(x^2 + x + 1)$.

1. **Identify the inner function and the outer function:**
   - Inner function: $u = x^2 + x + 1$
   - Outer function: $y = \sin(u)$

2. **Differentiate the inner function with respect to $x$:**

$$
\frac{du}{dx} = \frac{d}{dx} (x^2 + x + 1) = 2x + 1
$$

3. **Differentiate the outer function with respect to $u$:**

$$
\frac{dy}{du} = \frac{d}{du} \sin(u) = \cos(u)
$$

4. **Apply the chain rule:**

$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} = \cos(u) \cdot (2x + 1)
$$

5. **Substitute back $u = x^2 + x + 1$:**

$$
\frac{dy}{dx} = (2x + 1) \cdot \cos(x^2 + x + 1) 
$$

### Example 2: $\frac{d}{dx} \cos(x^3)$

Let's differentiate the function $y = \cos(x^3)$.

1. **Identify the inner function and the outer function:**
   - Inner function: $u = x^3$
   - Outer function: $y = \cos(u)$

2. **Differentiate the inner function with respect to $x$:**

$$
\frac{du}{dx} = \frac{d}{dx} (x^3) = 3x^2
$$

3. **Differentiate the outer function with respect to $u$:**

$$
\frac{dy}{du} = \frac{d}{du} \cos(u) = -\sin(u)
$$

4. **Apply the chain rule:**

$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} = -\sin(u) \cdot 3x^2
$$

5. **Substitute back $u = x^3$:**

$$
\frac{dy}{dx} = -3x^2 \cdot  \sin(x^3) 
$$
