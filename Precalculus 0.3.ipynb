# Families of Functions

Functions are often grouped into families based on their defining formulas or other common characteristics. Some of the most basic families of functions include constant functions, linear functions, and power functions.

## Constant Functions
The graph of a constant function $f(x) = c$ is a horizontal line on the Cartesian plane. Different values of $c$ generate a set of parallel horizontal lines.



## Linear Functions
Linear functions of the form $y = mx + b$ can be described in two families:

- When $b$ is fixed and $m$ varies, we get lines that intersect the y-axis at the same point but have different slopes.
- When $m$ is fixed and $b$ varies, we get a set of parallel lines with the same slope but different y-intercepts.



## Power Functions
The family of power functions $y = x^n$ changes based on the nature of $n$:

- For even $n$, the functions $f(x) = x^n$ are even, making their graphs symmetric about the y-axis. The graphs have the general shape of $y = x^2$ and pass through the points $(-1, 1)$, $(0, 0)$, and $(1, 1)$. As $n$ increases, the graphs flatten between $-1 < x < 1$ and steepen outside that interval.
- For odd $n$, the functions $f(x) = x^n$ are odd, resulting in graphs symmetric about the origin. The graphs have the general shape of $y = x^3$ and pass through the points $(-1, -1)$, $(0, 0)$, and $(1, 1)$. Similar to even powers, as $n$ increases, the flattening and steepening effects are observed.

The behavior of these functions can be explored further by considering the effect of raising a number $x$ to higher powers, especially noting the difference when $x$ is between $-1$ and $1$ versus outside that range.




# Inverse Proportional Functions and Their Graphs

Inverse proportional functions are of the form $y = \frac{k}{x}$, where $k$ is a constant. The graph of these functions exhibits characteristic hyperbolic shapes.

## Properties of Inverse Proportional Functions

- **Symmetry**: For $y = \frac{1}{x^n}$, if $n$ is even, the function is symmetric about the y-axis. If $n$ is odd, the function is symmetric about the origin.
- **Discontinuity**: There is a discontinuity at $x=0$ since division by zero is undefined.
- **Behavior at Origin**: As $x$ approaches zero, $y$ approaches infinity, which is described as a vertical asymptote at $x=0$.
- **Behavior at Infinity**: As $x$ approaches infinity, $y$ approaches zero, which is described as a horizontal asymptote at $y=0$.
- **Flattening and Steepening**: As $n$ increases, the graphs become steeper for $x$ values between $-1$ and $0$ or $0$ and $1$, and flatter for $|x| > 1$.



## The Families $y = \frac{1}{x^n}$

- **Even $n$**: For even values of $n$, the graph is always positive, and the function never takes negative values. The graph exhibits steepness near the origin and flattens out as $|x|$ grows.

- **Odd $n$**: For odd values of $n$, the graph passes through the points $(1, 1)$ and $(-1, -1)$ and exhibits symmetry about the origin, meaning the function is odd. The graph is in the first and third quadrants only.

## Example of Inverse Proportionality

Consider the experimental data suggesting that $y$ is inversely proportional to $x$ with a constant of proportionality $k = 5$:

$y = \frac{5}{x}$

This equation describes a hyperbola which, when graphed with the given data points, would pass through the points listed in the table, exhibiting inverse proportionality.

## Application to Boyle's Law

Boyles law states that for a given mass of an ideal gas at constant temperature, the product of pressure ($P$) and volume ($V$) is constant ($k$). This is an example of inverse proportionality:

$PV = k$ or $P = \frac{k}{V}$

Graphically, as the pressure increases, the volume decreases, creating a hyperbolic graph similar to the family $y = \frac{1}{x}$, but scaled according to the specific gas constant.

To analyze these functions mathematically, the concepts of limits and asymptotic behavior are crucial, especially as they relate to the undefined nature of division by zero and the behavior of the function as $x$ approaches zero or infinity. These properties are foundational in calculus and are used to understand a wide variety of physical phenomena.




# Power Functions with Noninteger Exponents

Power functions with noninteger exponents can have various forms and characteristics.

## Radical Functions

- **Square Root Function**: For $y = \sqrt{x}$, the graph is the upper half of the parabola $x = y^2$.
- **Cube Root Function**: For $y = \sqrt[3]{x}$, the domain is $(-\infty, +\infty)$, meaning the graph extends over the entire x-axis.

The general form of a radical function is $y = \sqrt[n]{x}$, where $n$ is the degree of the root.

- If $n$ is even, the graph only exists for $x \geq 0$.
- If $n$ is odd, the graph extends over all real numbers.

## Power Functions with Fractional Exponents

For functions like $f(x) = x^{2/3}$, the graph tends to flatten as $|x|$ grows and steepen near the origin. The behavior of these functions can be influenced by both the numerator and the denominator of the exponent.

## Graphing Considerations

Graphing utilities may not always display the entire graph of functions with fractional exponents correctly. Adjustments using absolute values may be necessary to graph $f(x) = x^{p/q}$ correctly.

For example:
- For even $p$ and odd $q$, graph $g(x) = |x|^{p/q}$.
- For odd $p$ and odd $q$, graph $g(x) = (|x|/x)|x|^{p/q}$.

## Polynomial Functions

Polynomials are sums of terms of the form $cx^n$ with a nonnegative integer $n$. The degree of the polynomial is the highest power of $x$ with a nonzero coefficient.

- **Linear**: Degree 1, $y = 2x + 1$
- **Quadratic**: Degree 2, $y = x^2 - 3x + 1$
- **Cubic**: Degree 3, $y = 2x^3 - 7$

Polynomial graphs are continuous and smooth without breaks or sharp corners.





## Rational Functions

Rational functions are ratios of polynomials, $f(x) = \frac{P(x)}{Q(x)}$. Their graphs can have discontinuities where $Q(x) = 0$ and may have horizontal and vertical asymptotes.


## Algebraic Functions

Algebraic functions are made from polynomials using algebraic operations, including root extraction. Their graphs can vary greatly, and calculus methods are often used for analysis.



---

The understanding of these functions and their graphs is essential in calculus, as they often model real-world phenomena and their rates of change.


# Families of Functions

Various families of functions such as power functions, polynomials, and trigonometric functions play significant roles in mathematics due to their unique properties and applications.

## Trigonometric Functions

Trigonometric functions of the form $f(x) = A \sin(Bx - C)$ and $g(x) = A \cos(Bx - C)$, where $A$, $B$, and $C$ are constants, are derived by transforming the base functions $\sin x$ and $\cos x$.


### Properties

- **Amplitude**: The maximum extent of a vibration or oscillation, measured from the position of equilibrium. Amplitude equals $|A|$.
- **Period**: The duration of one complete cycle of the function, calculated as $2\pi / |B|$.
- **Frequency**: The number of cycles the function completes in a unit interval, defined as $|B| / 2\pi$.


### Transformations

- Vertical stretching/compressing by a factor of $A$.
- Horizontal compressing/stretching by a factor of $1/B$.
- Horizontal shift to the right by $C/B$ units if $C/B > 0$ or to the left by $|C/B|$ units if $C/B < 0$.
- Reflection about the axes if $A$ or $B$ is negative.
