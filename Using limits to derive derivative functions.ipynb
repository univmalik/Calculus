# Calculating Derivatives of Linear Functions Using Limits

## Introduction

The derivative of a function $y = f(x)$, **denoted as $\frac{dy}{dx}$ or $f'(x)$**, represents the rate at which the function's value changes with respect to changes in $x$. For linear functions, this rate of change is constant. We can calculate the derivative using the limit definition.

## Linear Functions

A linear function is of the form:

$$
y = mx + b
$$

where:
- $m$ is the slope of the line.
- $b$ is the y-intercept, the point where the line crosses the y-axis.

### Derivative Using the Limit Definition

To find the derivative of $y = mx + b$ using the limit definition, we use:

$$
f'(x) = \lim_{{h \to 0}} \frac{f(x+h) - f(x)}{h}
$$

### Example: Derivative of $y = 3x + 2$

1. **Define the function**:
   $$
   f(x) = 3x + 2
   $$

2. **Apply the limit definition**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{f(x+h) - f(x)}{h}
   $$

3. **Calculate $f(x+h)$**:
   $$
   f(x+h) = 3(x+h) + 2 = 3x + 3h + 2
   $$

4. **Substitute $f(x+h)$ and $f(x)$ into the limit definition**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{(3x + 3h + 2) - (3x + 2)}{h}
   $$

5. **Simplify the expression**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{3x + 3h + 2 - 3x - 2}{h}
   $$
   $$
   f'(x) = \lim_{{h \to 0}} \frac{3h}{h}
   $$

6. **Cancel out $h$**:
   $$
   f'(x) = \lim_{{h \to 0}} 3 = 3
   $$

Thus, the derivative of $y = 3x + 2$ is $f'(x) = 3$, which is constant.

### General Case: Derivative of $y = mx + b$

Following the same steps for the general linear function $y = mx + b$:

1. **Define the function**:
   $$
   f(x) = mx + b
   $$

2. **Apply the limit definition**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{f(x+h) - f(x)}{h}
   $$

3. **Calculate $f(x+h)$**:
   $$
   f(x+h) = m(x+h) + b = mx + mh + b
   $$

4. **Substitute $f(x+h)$ and $f(x)$ into the limit definition**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{(mx + mh + b) - (mx + b)}{h}
   $$

5. **Simplify the expression**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{mx + mh + b - mx - b}{h}
   $$
   $$
   f'(x) = \lim_{{h \to 0}} \frac{mh}{h}
   $$

6. **Cancel out $h$**:
   $$
   f'(x) = \lim_{{h \to 0}} m = m
   $$

Thus, the derivative of any linear function $y = mx + b$ is $f'(x) = m$, which is constant.

## Conclusion

Using the limit definition of the derivative, we have shown that the derivative of a linear function $y = mx + b$ is simply the constant $m$. This indicates that the rate of change of $y$ with respect to $x$ is constant for linear functions. In the next section, we will explore the derivatives of exponential functions, where the rate of change is not constant.


# Calculating Derivatives Using Limits

## Introduction

The derivative of a function $y = f(x)$, denoted as $\frac{dy}{dx}$ or $f'(x)$, represents the rate at which the function's value changes with respect to changes in $x$. We can calculate the derivative using the limit definition.

## Derivative of $y = x^2$ Using Limits

To find the derivative of $y = x^2$ using the limit definition, we use:

$$
f'(x) = \lim_{{h \to 0}} \frac{f(x+h) - f(x)}{h}
$$

1. **Define the function**:
   $$
   f(x) = x^2
   $$

2. **Apply the limit definition**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{f(x+h) - f(x)}{h}
   $$

3. **Calculate $f(x+h)$**:
   $$
   f(x+h) = (x+h)^2 = x^2 + 2xh + h^2
   $$

4. **Substitute $f(x+h)$ and $f(x)$ into the limit definition**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{(x^2 + 2xh + h^2) - x^2}{h}
   $$

5. **Simplify the expression**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{x^2 + 2xh + h^2 - x^2}{h}
   $$
   $$
   f'(x) = \lim_{{h \to 0}} \frac{2xh + h^2}{h}
   $$
   $$
   f'(x) = \lim_{{h \to 0}} (2x + h)
   $$

6. **Take the limit as $h$ approaches 0**:
   $$
   f'(x) = 2x
   $$

Thus, the derivative of $y = x^2$ is $f'(x) = 2x$.

### Graph Explanation for $y = x^2$ and $y = 2x$

In the graph below, the red curve represents the function $y = x^2$ and the blue line represents its derivative $y = 2x$. The derivative function $y = 2x$ gives the slope of the tangent line to the curve at any point $x$ for $y = x^2$.



- The curve $y = x^2$ is a parabola opening upwards.
- The line $y = 2x$ is a straight line with a constant slope of 2. This means that for $y = x^2$, at every point the gradient of the curve is equal to $\frac{dy}{dx}=2x$ and is dependent upon $x$ and is not a constant.


## Derivative of $y = x^3$ Using Limits

To find the derivative of $y = x^3$ using the limit definition, we use:

$$
f'(x) = \lim_{{h \to 0}} \frac{f(x+h) - f(x)}{h}
$$

1. **Define the function**:
   $$
   f(x) = x^3
   $$

2. **Apply the limit definition**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{f(x+h) - f(x)}{h}
   $$

3. **Calculate $f(x+h)$**:
   $$
   f(x+h) = (x+h)^3 = x^3 + 3x^2h + 3xh^2 + h^3
   $$

4. **Substitute $f(x+h)$ and $f(x)$ into the limit definition**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{(x^3 + 3x^2h + 3xh^2 + h^3) - x^3}{h}
   $$

5. **Simplify the expression**:
   $$
   f'(x) = \lim_{{h \to 0}} \frac{x^3 + 3x^2h + 3xh^2 + h^3 - x^3}{h}
   $$
   $$
   f'(x) = \lim_{{h \to 0}} \frac{3x^2h + 3xh^2 + h^3}{h}
   $$
   $$
   f'(x) = \lim_{{h \to 0}} (3x^2 + 3xh + h^2)
   $$

6. **Take the limit as $h$ approaches 0**:
   $$
   f'(x) = 3x^2
   $$

Thus, the derivative of $y = x^3$ is $f'(x) = 3x^2$.

### Graph Explanation for $y = x^3$ and $y = 3x^2$

In the graph below, the red curve represents the function $y = x^3$ and the blue line represents its derivative $y = 3x^2$. The derivative function $y = 3x^2$ gives the slope of the tangent line to the curve at any point $x$.



- The curve $y = x^3$ is a cubic function that has points of inflection.

$$\frac{dy}{dx}=\frac{d[x^3]}{dx}=3x^2$$
- The curve $y = 3x^2$ is a parabola opening upwards, representing the changing slope of $y = x^3$.
- As you can see, the curve $y = x^3$ has a positive slope at all times even when $x<0$. And the slope function $y = 3x^2$ can never be negative.

## Conclusion

Using the limit definition of the derivative, we have shown how to calculate the derivatives of $y = x^2$ and $y = x^3$. These derivatives represent the rate of change of the functions with respect to $x$:
- For $y = x^2$, the derivative is $f'(x) = 2x$.
- For $y = x^3$, the derivative is $f'(x) = 3x^2$.

The graphs illustrate how the derivatives provide the slopes of the tangent lines to the original functions at any given point. In the next section, we will explore the derivatives of exponential functions.
