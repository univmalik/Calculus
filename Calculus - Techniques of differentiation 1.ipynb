# Introduction to Techniques of Differentiation

In calculus, differentiation is a fundamental operation that computes the rate at which a quantity changes. The derivative of a function at a point is the slope of the tangent line to the graph of the function at that point.

## Derivative of a Constant Function

Consider a constant function $f(x) = c$, where $c$ is a constant.



- **Graphical Representation**: The graph is a horizontal line intersecting the y-axis at $y=c$.
- **Slope of Tangent**: The slope of the tangent to the graph is zero for all $x$.
- **Derivative**: The derivative of a constant function is zero, mathematically represented as:

  $$
  \frac{d}{dx}[c] = 0
  $$

- **Derivation**: This is derived from the limit definition of a derivative as follows:

  $$
  f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} = \lim_{h \to 0} \frac{c - c}{h} = \lim_{h \to 0} 0 = 0
  $$
  
  

## Derivatives of Power Functions

Now, consider the simplest power function $f(x) = x$.

- **Graphical Representation**: The graph is a line through the origin with a slope of 1.
- **Slope of Tangent**: The slope of the tangent to the graph is one for all $x$.

- **Derivative**: The derivative of $f(x) = x$ is:

  $$
  \frac{d}{dx}[x] = 1
  $$

- **Power Rule**: A general rule for power functions $f(x) = x^n$ is:

  $$
  \frac{d}{dx}[x^n] = nx^{n-1}
  $$

  where $n$ is a positive integer.

- **Extended Power Rule**: The power rule is extended to any real number exponent $r$:

  $$
  \frac{d}{dx}[x^r] = rx^{r-1}
  $$

  This implies that to differentiate a power function, decrease the exponent by one and multiply by the original exponent.

## Examples

- The derivative of $x$ raised to the power of $\frac{1}{2}$:

  $$
  \frac{d}{dx}[x^{1/2}] = \frac{1}{2}x^{-1/2}
  $$

- The derivative of $x$ raised to the power of $-1$:

  $$
  \frac{d}{dx}[x^{-1}] = -x^{-2}
  $$

These rules and examples form the foundation of differentiating power functions without resorting to the limit definition for each case.


# Derivative of a Constant Times a Function

Calculus provides us with rules to differentiate functions efficiently. One such rule is for differentiating a constant multiplied by a function.

## Constant Multiple Rule

**Theorem 2.3.4 (Constant Multiple Rule):** If $f$ is differentiable at $x$ and $c$ is any real number, then the product $cf$ is also differentiable at $x$, and the derivative is given by:

$$
\frac{d}{dx}[cf(x)] = c\frac{d}{dx}[f(x)]
$$

This can be understood by taking the derivative of $cf(x)$, which is the product of a constant $c$ and a function $f(x)$:

$$
\frac{d}{dx}[cf(x)] = \lim_{h \to 0} \frac{cf(x + h) - cf(x)}{h} = c\lim_{h \to 0} \frac{f(x + h) - f(x)}{h} = c\frac{d}{dx}[f(x)]
$$

A constant factor can be moved outside the limit operation, simplifying the differentiation process.

**Example:**

$$
\frac{d}{dx}[4x^8] = 4\frac{d}{dx}[x^8] = 4[8x^7] = 32x^7
$$

## Derivatives of Sums and Differences

The derivatives of sums and differences are as straightforward as the derivatives of constant multiples.

**Theorem 2.3.5 (Sum and Difference Rules):** If $f$ and $g$ are differentiable at $x$, then so are $f + g$ and $f - g$, and their derivatives are:

$$
\frac{d}{dx}[f(x) + g(x)] = \frac{d}{dx}[f(x)] + \frac{d}{dx}[g(x)]
$$

$$
\frac{d}{dx}[f(x) - g(x)] = \frac{d}{dx}[f(x)] - \frac{d}{dx}[g(x)]
$$

**Example:**

$$
\frac{d}{dx}[2x^6 + x^{-9}] = \frac{d}{dx}[2x^6] + \frac{d}{dx}[x^{-9}] = 12x^5 - 9x^{-10}
$$

These rules can be applied to differentiate any polynomial by treating each term separately.

## Application in Differentiating Polynomials

By utilizing the constant multiple rule and the sum and difference rules, we can differentiate complex polynomials.

**Example:**

Find $\frac{dy}{dx}$ if $y = 3x^8 - 2x^5 + 6x + 1$.

**Solution:**

$$
\frac{dy}{dx} = \frac{d}{dx}[3x^8] - \frac{d}{dx}[2x^5] + \frac{d}{dx}[6x] + \frac{d}{dx}[1] = 24x^7 - 10x^4 + 6 $$


# Horizontal Tangents and Areas of Triangles

## Example 7: Horizontal Tangent Lines

**Problem:** Find the points where the graph of $y = x^3 - 3x + 4$ has horizontal tangent lines.

**Solution:**
A horizontal tangent occurs where the slope of the curve is zero. To find it, we differentiate the function and set the derivative equal to zero.

   

1. Compute the derivative of the function:
   $$
   y'(x) = \frac{d}{dx}(x^3 - 3x + 4) = 3x^2 - 3
   $$

2. Set the derivative equal to zero to find where the slope is zero:
   $$
   3x^2 - 3 = 0
   $$

3. Solve for $x$:
   $$
   x^2 - 1 = 0 \quad \Rightarrow \quad (x - 1)(x + 1) = 0 \quad \Rightarrow \quad x = -1 \text{ or } x = 1
   $$

4. Find the corresponding points on the graph:
   For $x = -1$:
   $$
   y(-1) = (-1)^3 - 3(-1) + 4 = 6
   $$
   For $x = 1$:
   $$
   y(1) = 1^3 - 3(1) + 4 = 2
   $$
   Thus, the points are $(-1, 6)$ and $(1, 2)$.

## Example 8: Area of a Triangle

**Problem:** Determine the area of the triangle formed by the coordinate axes and the tangent line to the curve $y = 5x^{-1} - \frac{1}{5}x$ at the point $(5, 0)$.


**Solution:**
1. Differentiate the function to find the slope of the tangent:
   $$
   y'(x) = \frac{d}{dx}\left(5x^{-1} - \frac{1}{5}x\right) = -5x^{-2} - \frac{1}{5}
   $$

2. Evaluate the derivative at $x = 5$ to get the slope of the tangent line at that point:
   $$
   y'(5) = -\frac{5}{5^2} - \frac{1}{5} = -\frac{1}{5} - \frac{1}{5} = -\frac{2}{5}
   $$

3. Write the equation of the tangent line using the point-slope form:
   $$
   y - 0 = -\frac{2}{5}(x - 5) \quad \Rightarrow \quad y = -\frac{2}{5}x + 2
   $$

4. Calculate the area of the triangle with base 5 and height 2:
   $$
   \text{Area} = \frac{1}{2} \times \text{base} \times \text{height} = \frac{1}{2} \times 5 \times 2 = 5
   $$

# Higher Derivatives

The concept of higher derivatives is an extension of the basic differentiation process. After obtaining the first derivative of a function $f$, we can differentiate it further if the resulting function is also differentiable.

## Definition

The second derivative of $f$ is denoted by $f''$ and is the derivative of the first derivative $f'$. This process can continue to yield third, fourth, fifth, and higher-order derivatives:

- Second derivative: $f'' = (f')'$
- Third derivative: $f''' = (f'')'$
- Fourth derivative: $f^{(4)} = (f''')'$
- Fifth derivative: $f^{(5)} = (f^{(4)})'$
- And so on...

If we have $y = f(x)$, then the derivatives can also be represented as $y'$, $y''$, $y'''$, $y^{(4)}$, $y^{(5)}$, and so forth.

## Notation

The derivatives of a function are commonly denoted in several ways:

- First derivative: $y' = \frac{dy}{dx} = \frac{d}{dx}[f(x)]$
- Second derivative: $y'' = \frac{d^2y}{dx^2} = \frac{d}{dx}\left(\frac{d}{dx}[f(x)]\right) = \frac{d^2}{dx^2}[f(x)]$
- Third derivative: $y''' = \frac{d^3y}{dx^3} = \frac{d}{dx}\left(\frac{d^2}{dx^2}[f(x)]\right) = \frac{d^3}{dx^3}[f(x)]$
- And similarly for higher-order derivatives.

The nth order derivative is generally denoted as:

$$
\frac{d^n y}{dx^n} = f^{(n)}(x) = \frac{d^n}{dx^n}[f(x)]
$$

And the value of the nth order derivative at a specific point $x = x_0$ is:

$$
\left.\frac{d^n y}{dx^n}\right|_{x=x_0} = f^{(n)}(x_0) = \left.\frac{d^n}{dx^n}[f(x)]\right|_{x=x_0}
$$

## Example 9: Calculating Higher Derivatives

Given the function $f(x) = 3x^4 - 2x^3 + x^2 - 4x + 2$, the higher derivatives are computed as follows:

- First derivative: $f'(x) = 12x^3 - 6x^2 + 2x - 4$
- Second derivative: $f''(x) = 36x^2 - 12x + 2$
- Third derivative: $f'''(x) = 72x - 12$
- Fourth derivative: $f^{(4)}(x) = 72$
- Fifth derivative: $f^{(5)}(x) = 0$
- And for all $n \geq 5$, $f^{(n)}(x) = 0$

The significance of second and higher-order derivatives will be explored in later sections of calculus.


# Differentiation Examples

We will differentiate two functions by rewriting them as powers of $x$ and then applying the Power Rule.

## Example (a)

Differentiate $f(x) = \frac{-1}{x^2}$.

**Solution:**

First, rewrite the function to use a negative exponent:
$$
f(x) = -x^{-2}
$$

Now, apply the Power Rule ($\frac{d}{dx}[x^n] = nx^{n-1}$) with $n = -2$:
$$
f'(x) = \frac{d}{dx}[-x^{-2}] = -2x^{-2-1} = -2x^{-3}
$$

Simplify the expression to get the final result:
$$
f'(x) = -\frac{2}{x^3}
$$

## Example (b)

Differentiate $y = (3x^2)^{1/3}$.

**Solution:**

We will apply the Power Rule to the function expressed as a power of $x$. The exponent $1/3$ indicates a cube root.

Rewrite the function:
$$
y = (3x^2)^{1/3}
$$

Differentiate using the Power Rule:
$$
\frac{dy}{dx} = \frac{d}{dx}[(3x^2)^{1/3}] = \frac{1}{3}(3x^2)^{-2/3} \cdot 2 \cdot 3x = 2x(3x^2)^{-2/3}
$$

Simplify the derivative:
$$
\frac{dy}{dx} = 2x \cdot 3^{-2/3} \cdot x^{-4/3} = \frac{2}{3^{2/3}} \cdot x^{-1/3}
$$

The final derivative is:
$$
\frac{dy}{dx} = \frac{2}{3^{2/3}x^{1/3}}
$$

In both examples, we have applied the Power Rule to differentiate functions that are expressed in terms of a power of $x$. The negative exponent in the first example and the fractional exponent in the second example are handled accordingly to arrive at the respective derivatives.


# Tangent and Normal Lines to the Curve

We will find the equations of the tangent and normal lines to the curve $y = x\sqrt{x}$ at the point $(1, 1)$.

## Finding the Derivative

The derivative of $f(x) = x\sqrt{x}$ can be found using the power rule and chain rule.

$$
f(x) = x \cdot x^{1/2}
$$

Combine terms to make differentiation easier:

$$
f(x) = x^{3/2}
$$

Differentiate using the power rule:

$$
f'(x) = \frac{3}{2}x^{3/2 - 1} = \frac{3}{2}x^{1/2}
$$

So the slope of the tangent line at the point $(1, 1)$ is:

$$
f'(1) = \frac{3}{2}
$$

## Equation of the Tangent Line

The equation of the tangent line at the point $(1, 1)$ using the point-slope form is:

$$
y - 1 = \frac{3}{2}(x - 1)
$$

Simplify the equation:

$$
y = \frac{3}{2}x - \frac{1}{2} - 1
$$

$$
y = \frac{3}{2}x - \frac{1}{2}
$$

## Equation of the Normal Line

The normal line has a slope that is the negative reciprocal of the tangent line's slope. Therefore, the slope of the normal line is: $-\frac{2}{3}$

Using the point-slope form again for the normal line:

$$
y - 1 = -\frac{2}{3}(x - 1)
$$

Simplify the equation:

$$
y = -\frac{2}{3}x + \frac{2}{3} + 1
$$

$$
y = -\frac{2}{3}x + \frac{5}{3}
$$

## Graphing

Plot the curve $y = x\sqrt{x}$ and the tangent and normal lines at the point $(1, 1)$. The tangent line has a slope of $\frac{3}{2}$, and the normal line has a slope of $-\frac{2}{3}$.

# Calculus Problems: Horizontal Tangents and Particle Acceleration

## Horizontal Tangents

We want to find the points on the curve $y = x^4 - 6x^2 + 4$ where the tangent line is horizontal.

**Solution:**

A horizontal tangent occurs where the derivative of the curve with respect to $x$ is zero. So, we differentiate $y$ with respect to $x$:

$$
\frac{dy}{dx} = \frac{d}{dx}(x^4) - 2 \cdot 6 \frac{d}{dx}(x^2) + \frac{d}{dx}(4)
$$

Applying the power rule:

$$
\frac{dy}{dx} = 4x^3 - 12x
$$

To find horizontal tangents, we set the derivative equal to zero:

$$
4x^3 - 12x = 0
$$

Factor out the common term:

$$
4x(x^2 - 3) = 0
$$

Solve for $x$:

$$
x = 0 \quad \text{or} \quad x^2 = 3 \quad \Rightarrow \quad x = \pm \sqrt{3}
$$

The points of horizontal tangents are at $x = 0$, $x = \sqrt{3}$, and $x = -\sqrt{3}$. We plug these back into the original equation to find the corresponding $y$ values:

$$
y(0) = 0^4 - 6 \cdot 0^2 + 4 = 4
$$

$$
y(\sqrt{3}) = (\sqrt{3})^4 - 6(\sqrt{3})^2 + 4 = 9 - 18 + 4 = -5
$$

$$
y(-\sqrt{3}) = (-\sqrt{3})^4 - 6(-\sqrt{3})^2 + 4 = 9 - 18 + 4 = -5
$$

Thus, the curve has horizontal tangents at the points $(0, 4)$, $(\sqrt{3}, -5)$, and $(-\sqrt{3}, -5)$.

## Particle Acceleration

Given the equation of motion $s = 2t^3 - 5t^2 + 3t + 4$, where $s$ is in centimeters and $t$ in seconds, we need to find the acceleration as a function of time, and specifically after 2 seconds.

**Solution:**

The velocity $v(t)$ is the first derivative of $s$ with respect to $t$:

$$
v(t) = \frac{ds}{dt} = 6t^2 - 10t + 3
$$

The acceleration $a(t)$ is the derivative of the velocity:

$$
a(t) = \frac{dv}{dt} = 12t - 10
$$

After 2 seconds, the acceleration is:

$$
a(2) = 12 \cdot 2 - 10 = 14 \text{ cm/s}^2 $$




## Problem 4

**Given Function:**
$y = \frac{1}{2}(x^4 + 7)$

**To find:** $\frac{dy}{dx}$

**Solution:**

1. Apply the constant multiple rule:
   The derivative of a constant times a function is the constant times the derivative of the function. So, we can pull out the $\frac{1}{2}$ and differentiate the inside function.

   $$
   \frac{dy}{dx} = \frac{1}{2} \cdot \frac{d}{dx}(x^4 + 7)
   $$

2. Use the sum rule:
   The derivative of a sum is the sum of the derivatives.

   $$
   \frac{dy}{dx} = \frac{1}{2} \left(\frac{d}{dx}(x^4) + \frac{d}{dx}(7)\right)
   $$

3. Apply the power rule and the derivative of a constant:
   The power rule states that $\frac{d}{dx}(x^n) = nx^{n-1}$, and the derivative of a constant is 0.

   $$
   \frac{dy}{dx} = \frac{1}{2} \left(4x^3 + 0\right)
   $$

4. Simplify the expression to get the final derivative:

   $$
   \frac{dy}{dx} = 2x^3
   $$

**The derivative of $y = \frac{1}{2}(x^4 + 7)$ is $2x^3$.**

## Additional Example

**Given Function:**
$f(x) = (3x^2 + 1)^2$

**To find:** $f'(x)$

**Solution:**

1. Expand the function using the binomial theorem:
   
   $$
   f(x) = 9x^4 + 6x^2 + 1
   $$

2. Differentiate each term separately using the sum rule and constant multiple rule:

   $$
   f'(x) = \frac{d}{dx}(9x^4) + \frac{d}{dx}(6x^2) + \frac{d}{dx}(1)
   $$

3. Apply the power rule to each term:

   $$
   f'(x) = 36x^3 + 12x
   $$

4. Since the derivative of a constant is 0, we can ignore the last term after differentiation.

**The derivative of $f(x) = (3x^2 + 1)^2$ is $f'(x) = 36x^3 + 12x$.**


# Instantaneous Rate of Change for a Spherical Balloon

## Part (a): General Formula for Rate of Change

We're given the volume of a sphere as:

$$
V = \frac{4}{3}\pi r^3
$$

To find the instantaneous rate of change of the volume $V$ with respect to the radius $r$, we need to differentiate $V$ with respect to $r$.

**Solution:**

1. Using the power rule for differentiation:

$$
\frac{dV}{dr} = \frac{d}{dr}\left(\frac{4}{3}\pi r^3\right)
$$

2. Apply the constant multiple rule and power rule:

$$
\frac{dV}{dr} = \frac{4}{3} \pi \cdot 3r^{3-1} = 4\pi r^2
$$

We observe that $\frac{dV}{dr}$ is equivalent to the surface area of the sphere.

**Result:**

The general formula for the instantaneous rate of change of the volume with respect to the radius is:

$$
\frac{dV}{dr} = 4\pi r^2
$$

## Part (b): Rate of Change When $r = 5$

Now, we will calculate the rate of change of $V$ with respect to $r$ when the radius is $5$ cm.

**Solution:**

1. We start with the formula derived in part (a):

$$
\frac{dV}{dr} = 4\pi r^2
$$

2. Substitute $r = 5$ into the formula:

$$
\left.\frac{dV}{dr}\right|_{r=5} = 4\pi (5)^2 = 100\pi
$$

**Result:**

The rate of change of the volume with respect to the radius when $r = 5$ is:

$$
\left.\frac{dV}{dr}\right|_{r=5} = 100\pi \text{ cm}^3/\text{cm}
