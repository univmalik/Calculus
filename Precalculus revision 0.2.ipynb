# Function Composition and Its Properties

Function composition is a fundamental concept in mathematics, particularly in the study of functions. It involves creating new functions by applying one function to the results of another.

## Basic Concepts

### Definition
Given two functions $f(x)$ and $g(x)$, the composition of $f$ and $g$, denoted as $(f \circ g)(x)$, is defined as $f(g(x))$.

### Properties
- **Non-Commutativity**: Generally, $(f \circ g)(x) \neq (g \circ f)(x)$.
- **Domain Considerations**: The domain of $(f \circ g)(x)$ is determined by the domain of $g$ and the range of $g$ that lies within the domain of $f$.

## Examples

### Example 1
Let $f(x) = x^2 + 3$ and $g(x) = \sqrt{x}$.

#### Part (a)
Find $(f \circ g)(x)$.

**Solution**:

$$
(f \circ g)(x) = f(g(x)) = [g(x)]^2 + 3 = (\sqrt{x})^2 + 3 = x + 3
$$

- Domain: Since $g(x)$ is $\sqrt{x}$, its domain is $[0, \infty)$. Thus, the domain of $(f \circ g)$ is also $[0, \infty)$.

#### Part (b)
Find $(g \circ f)(x)$.

**Solution**:

$$
(g \circ f)(x) = g(f(x)) = \sqrt{f(x)} = \sqrt{x^2 + 3}
$$

- Domain: The domain of $f(x)$ is $(-\infty, \infty)$, and since $x^2 + 3 \geq 0$ for all $x$, the domain of $(g \circ f)(x)$ is also $(-\infty, \infty)$.

### Example 2
Express $h(x) = (x + 1)^2$ as a composition of simpler functions.

**Solution**:

- Consider $g(x) = x + 1$ and $f(x) = x^2$.
- Then, $h(x) = [g(x)]^2 = f(g(x))$.
- Thus, $h = f \circ g$.

### Example 3
Express $\sin(x^3)$ as a composition of two functions.

**Solution**:

- Let $g(x) = x^3$ (inside function) and $f(x) = \sin(x)$ (outside function).
- Then, $\sin(x^3) = f(g(x))$.

## Decomposing Functions into Compositions

- A function can often be expressed as a composition of two or more simpler functions.
- The process involves identifying an "inside function" and an "outside function".
- This method is particularly useful in calculus and algebra for simplifying complex functions.

## Table of Function Compositions

| Function | Composition | Inside ($g(x)$) | Outside ($f(x)$) |
|----------|-------------|------------------|------------------|
| $(x^2 + 1)^{10}$ | $f(g(x))$ | $x^2 + 1$ | $x^{10}$ |
| $\sin^3(x)$ | $f(g(x))$ | $\sin(x)$ | $x^3$ |
| $\tan(x^5)$ | $f(g(x))$ | $x^5$ | $\tan(x)$ |
| $\sqrt{4 - 3x}$ | $f(g(x))$ | $4 - 3x$ | $\sqrt{x}$ |
| $8 + \sqrt{x}$ | $f(g(x))$ | $\sqrt{x}$ | $8 + x$ |

**Note**: Expressing functions as compositions can vary. For example, $(x^2 + 1)^{10}$ can also be expressed differently as shown in the given solutions.

---

In conclusion, understanding function composition is essential in mathematical analysis, allowing for the simplification and manipulation of complex expressions.


# Graphical Transformations of Functions

Graphical transformations allow us to use basic function graphs to understand and construct graphs of more complex functions. By applying simple operations such as addition, subtraction, and composition, we can visualize the changes to the graph of a function.

## Composition of Functions

Consider the function $y = \sqrt{x} + \frac{1}{x}$. To sketch this function, we can use the individual graphs of $y = \sqrt{x}$ and $y = \frac{1}{x}$, and add the corresponding $y$-coordinates for each $x \geq 0$.



### Example

To illustrate this with an example, if we want to sketch the graph of the function $y = \sqrt{x} - \frac{1}{x}$, we would perform the following steps:

1. Plot the graph of $y = \sqrt{x}$.
2. Plot the graph of $y = \frac{1}{x}$.
3. For each $x \geq 0$, subtract the $y$-value of $\frac{1}{x}$ from the $y$-value of $\sqrt{x}$.
4. The resulting points give the graph of $y = \sqrt{x} - \frac{1}{x}$.

## Translations of Functions

Translating a function involves shifting the graph horizontally and/or vertically. This can be done by adding or subtracting a constant to the function's output (vertical shift) or its input (horizontal shift).

### Table of Translations

| Operation on $y = f(x)$ | New Equation | Geometric Effect |
|-------------------------|--------------|------------------|
| Add a positive constant $c$ to $f(x)$ | $y = f(x) + c$ | Shifts the graph up by $c$ units |
| Subtract a positive constant $c$ from $f(x)$ | $y = f(x) - c$ | Shifts the graph down by $c$ units |
| Add a positive constant $c$ to $x$ | $y = f(x + c)$ | Shifts the graph left by $c$ units |
| Subtract a positive constant $c$ from $x$ | $y = f(x - c)$ | Shifts the graph right by $c$ units |



### Examples

1. To sketch $y = \sqrt{x} - 3$, shift the graph of $y = \sqrt{x}$ to the right by 3 units.
2. For $y = \sqrt{x} + 3$, shift the graph of $y = \sqrt{x}$ to the left by 3 units.
3. The graph of $y = x^2 - 4x + 5$ can be obtained by completing the square to write the function as $y = (x - 2)^2 + 1$ and then shifting the graph of $y = x^2$ right by 2 units and up by 1 unit.



By using these principles, we can construct the graphs of many functions from the basic graphs of $y = x^2$, $y = \sqrt{x}$, and $y = \frac{1}{x}$.





# Function Transformations and Symmetry

Function transformations include operations such as reflections, translations, stretches, and compressions, which modify the graph of a function in various ways.

## Reflections

- **Reflection about the y-axis**: If you replace $x$ with $-x$ in the function $f(x)$, the new function $f(-x)$ will reflect the graph of $f(x)$ about the y-axis.
- **Reflection about the x-axis**: If you multiply the function $f(x)$ by $-1$, the new function $-f(x)$ will reflect the graph of $f(x)$ about the x-axis.





## Translations

- **Vertical Translations**: Adding or subtracting a constant from $f(x)$ shifts the graph vertically.
- **Horizontal Translations**: Adding or subtracting a constant from $x$ shifts the graph horizontally.



## Stretches and Compressions

- **Vertical Stretches and Compressions**: Multiplying $f(x)$ by a constant $c$ stretches or compresses the graph vertically depending on whether $c > 1$ or $0 < c < 1$.
- **Horizontal Stretches and Compressions**: Multiplying $x$ by a constant $c$ compresses or stretches the graph horizontally depending on whether $c > 1$ or $0 < c < 1$.



## Symmetry

- **Symmetry about the x-axis**: A function is symmetric about the x-axis if for every point $(x, y)$, the point $(x, -y)$ is also on the graph.
- **Symmetry about the y-axis**: A function is symmetric about the y-axis if for every point $(x, y)$, the point $(-x, y)$ is also on the graph.
- **Symmetry about the origin**: A function is symmetric about the origin if for every point $(x, y)$, the point $(-x, -y)$ is also on the graph.



## Even and Odd Functions

- **Even Functions**: A function $f$ is even if $f(-x) = f(x)$. Its graph is symmetric about the y-axis.
- **Odd Functions**: A function $f$ is odd if $f(-x) = -f(x)$. Its graph is symmetric about the origin.



### Examples

1. To sketch $y = 3\sqrt{2 - x}$:
   - Reflect $y = 3\sqrt{x}$ about the y-axis to get $y = 3\sqrt{-x}$.
   - Then translate it right by 2 units.

   

2. To sketch $y = 4 - |x - 2|$:
   - Translate $y = |x|$ right by 2 units to get $y = |x - 2|$.
   - Reflect it about the x-axis and then translate it up by 4 units.

These transformations and symmetries are powerful tools for analyzing and sketching the graphs of complex functions based on simple ones.
