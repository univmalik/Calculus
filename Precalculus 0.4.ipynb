# Inverse Functions and Inverse Trigonometric Functions

In mathematics, the concept of an inverse is crucial as it refers to functions that reverse each others effects. This section will explore inverse functions, especially as they pertain to trigonometric functions, which allow us to recover an angle from a given trigonometric function value.

## Inverse Functions

The process of solving an equation $y = f(x)$ for $x$ in terms of $y$, denoted as $x = g(y)$, is a fundamental mathematical principle. For instance, the equation 

$$ y = x^3 + 1 $$

can be inverted to solve for $x$:

$$ x = \sqrt[3]{y - 1} $$

Here, the original equation is better suited when $x$ is known and we want to find $y$, and the inverted equation is used when $y$ is known and we need to determine $x$ (refer to Figure 0.4.1).

The functions $f(x) = x^3 + 1$ and $g(y) = \sqrt[3]{y - 1}$, when composed, negate each other such that:

$$ g(f(x)) = \sqrt[3]{f(x) - 1} = \sqrt[3]{x^3 + 1 - 1} = x $$
$$ f(g(y)) = [g(y)]^3 + 1 = (\sqrt[3]{y - 1})^3 + 1 = y $$



## Definition of Inverse Functions

If functions $f$ and $g$ satisfy:

$$ g(f(x)) = x \quad \text{for every } x \text{ in the domain of } f $$
$$ f(g(y)) = y \quad \text{for every } y \text{ in the domain of } g $$

then $f$ is an inverse of $g$ and $g$ is an inverse of $f$. We denote the inverse of $f$ by $f^{-1}$.

It is proven that if a function $f$ has an inverse, this inverse is unique, which allows us to refer to "the" inverse of $f$ with the notation $f^{-1}$.

## Graphing Inverse Functions

To graph $f$ and $f^{-1}$ on the same coordinate system, we use the same independent variable for both. For the functions $f(x) = x^3 + 1$ and $f^{-1}(y) = \sqrt[3]{y - 1}$, we would express them with $x$ as the independent variable and graph:

$$ y = x^3 + 1 $$
$$ y = \sqrt[3]{x - 1} $$

These are known as the cancellation equations for $f$ and $f^{-1}$.

## Existence of Inverse Functions

A function $f$ has an inverse if and only if it is one-to-one, which means that it assigns distinct outputs to distinct inputs. Geometrically, this can be tested by ensuring that any horizontal line cuts the graph of $y = f(x)$ at most once, known as the Horizontal Line Test.

The existence of inverse functions is not always guaranteed and can fail if the equation $y = f(x)$ cannot be explicitly solved for $x$ as a function of $y$.




# Inverse Functions and Inverse Trigonometric Functions

The term "inverse" in mathematics describes functions that reverse each other's effects. Inverse trigonometric functions, specifically, allow us to recover an angle from a given trigonometric function value.

## Inverse Functions

Solving an equation $y = f(x)$ for $x$ as a function of $y$, denoted as $x = g(y)$, is a foundational concept. For example, the equation

$$ y = x^3 + 1 $$

can be solved for $x$:

$$ x = \sqrt[3]{y - 1} $$

The original equation is preferable when $x$ is known and we want to find $y$, and the inverse equation is used when $y$ is known to determine $x$ (See Figure 0.4.1).



## Definition of Inverse Functions

If functions $f$ and $g$ satisfy the conditions:

$$ g(f(x)) = x \quad \forall x \in \text{domain of } f $$
$$ f(g(y)) = y \quad \forall y \in \text{domain of } g $$

then $f$ is an inverse of $g$ and $g$ is an inverse of $f$, and they are called inverse functions. The inverse of $f$ is denoted by $f^{-1}$.

## Changing the Independent Variable

While it's often convenient to use different independent variables for $f$ and $f^{-1}$, there are times when the same independent variable is preferable. For graphing $f$ and $f^{-1}$ in the same coordinate system, we would use $x$ as the independent variable and $y$ as the dependent variable for both. This leads to graphing the equations:

$$ y = x^3 + 1 $$
$$ y = \sqrt[3]{x - 1} $$

## Finding Inverse Functions

A method for finding the inverse of a function $f$ is as follows:

1. Write down the equation $y = f(x)$.
2. Solve this equation for $x$ as a function of $y$.
3. The resulting equation will be $x = f^{-1}(y)$.
4. If the same independent variable is desired for both $f$ and $f^{-1}$, interchange $x$ and $y$ in the equation to obtain $y = f^{-1}(x)$.

For example, to find the inverse of $f(x) = \sqrt{3x - 2}$, we write $y = \sqrt{3x - 2}$ and solve for $x$:

$$ y^2 = 3x - 2 $$
$$ x = \frac{1}{3}(y^2 + 2) $$

This gives us $f^{-1}(y) = \frac{1}{3}(y^2 + 2)$, which can be written with $x$ as the independent variable as:

$$ f^{-1}(x) = \frac{1}{3}(x^2 + 2), \quad x \geq 0 $$

The domain of $f^{-1}$ is the range of $f$, and the range of $f^{-1}$ is the domain of $f$.



## Existence of Inverse Functions

A function $f$ has an inverse if and only if it is one-to-one. A geometric test for this is the Horizontal Line Test, which states that a function has an inverse if and only if any horizontal line cuts the graph of $y = f(x)$ at most once.


# Inverse Functions and Inverse Trigonometric Functions

In mathematics, the term *inverse* is used to describe functions that reverse one another in the sense that each undoes the effect of the other.

## Inverse Functions

The process of solving an equation $y = f(x)$ for $x$ in terms of $y$, denoted as $x = g(y)$, is central in mathematics. For instance, the equation $y = x^3 + 1$ can be solved for $x$ as:

$$ x = \sqrt[3]{y - 1} $$

### Definition

If functions $f$ and $g$ satisfy:

$$ g(f(x)) = x \text{ for every } x \text{ in the domain of } f $$
$$ f(g(y)) = y \text{ for every } y \text{ in the domain of } g $$

then $f$ is an inverse of $g$ and $g$ is an inverse of $f$, and they are called inverse functions.

It is noted that if a function $f$ has an inverse, that inverse is unique and is denoted as $f^{-1}$.

### Cancellation Equations

The equations that express the relationship between $f$ and $f^{-1}$ are:

$$ f^{-1}(f(x)) = x \text{ for every } x \text{ in the domain of } f $$
$$ f(f^{-1}(y)) = y \text{ for every } y \text{ in the domain of } f^{-1} $$

### Changing the Independent Variable

To graph $f$ and $f^{-1}$ on the same coordinate system, we may need to use the same independent variable for both. This leads to the reformulation of the cancellation equations using $x$ for both $f$ and $f^{-1}$.

## Domain and Range of Inverse Functions

The domain of $f^{-1}$ is the range of $f$, and the range of $f^{-1}$ is the domain of $f$.

## Finding Inverse Functions

To find the inverse of a function $f$:

1. Write $y = f(x)$.
2. Solve for $x$ as a function of $y$.
3. The solution is $x = f^{-1}(y)$.
4. If needed, interchange $x$ and $y$ to have $y = f^{-1}(x)$ with $x$ as the independent variable.



## Existence of Inverse Functions

A function $f$ has an inverse if and only if it is one-to-one, meaning $f(x_1) \neq f(x_2)$ whenever $x_1 \neq x_2$. This is visually represented by the graph of $y = f(x)$ being cut at most once by any horizontal line.

### The Horizontal Line Test

A function has an inverse if and only if its graph passes the horizontal line test, which means it is cut at most once by any horizontal line.



## Graphs of Inverse Functions

If $f$ has an inverse, then the graphs of $y = f(x)$ and $y = f^{-1}(x)$ are reflections of one another about the line $y = x$.

# Restricting Domains for Invertibility

If a function $g$ is obtained from a function $f$ by placing restrictions on the domain of $f$, then $g$ is called a restriction of $f$. For example, the function $g(x) = x^3, x \geq 0$ is a restriction of the function $f(x) = x^3$. More precisely, it is called the restriction of $x^3$ to the interval $[0, \infty)$.

It is possible to create an invertible function from a function that is not invertible by restricting the domain appropriately. For instance, the function $f(x) = x^2$ is not invertible over its entire domain. However, by considering the restricted functions

$$ f_1(x) = x^2, x \geq 0 $$
$$ f_2(x) = x^2, x \leq 0 $$

which together make up the complete graph of $f(x) = x^2$, we obtain functions that are each one-to-one and hence invertible. Their graphs pass the horizontal line test.

The inverses of these restricted functions are given by:

$$ f^{-1}_1(x) = \sqrt{x} $$
$$ f^{-1}_2(x) = -\sqrt{x} $$

These inverses correspond to the positive and negative square roots, respectively, and they reflect the one-to-one nature of the restricted domains.

Here's how the functions and their inverses are graphed:

- For $f_1(x) = x^2$ with $x \geq 0$, the inverse is $f^{-1}_1(x) = \sqrt{x}$.
- For $f_2(x) = x^2$ with $x \leq 0$, the inverse is $f^{-1}_2(x) = -\sqrt{x}$.

The graph of $y = x^2$ for $x \geq 0$ and $x \leq 0$, along with the graphs of $y = \sqrt{x}$ and $y = -\sqrt{x}$, demonstrate the concept of restricting the domain to make a function invertible.
